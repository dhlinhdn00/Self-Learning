# **SOLUTION**

```python
class Solution(object):
    def maxFrequency(self, nums, k, numOperations):
        nums.sort()
        n = len(nums)
        if n == 0:
            return 0
        if n == 1:
            return 1

        l = 0
        max_cover = 1
        for r in range(n):
            while nums[r] - nums[l] > 2 * k:
                l += 1
            max_cover = max(max_cover, r - l + 1)
        best_any_point = min(max_cover, numOperations)  # 

        ans = best_any_point
        i = 0
        L = 0
        R = 0
        while i < n:
            v = nums[i]
            j = i
            while j + 1 < n and nums[j + 1] == v:
                j += 1
            freq_v = j - i + 1

            while R < n and nums[R] <= v + k:
                R += 1
            while L < n and nums[L] < v - k:
                L += 1

            C_v = R - L
            cand_v = min(C_v, freq_v + numOperations)
            if cand_v > ans:
                ans = cand_v

            i = j + 1

        return min(ans, n)
```

# **DRAFT**

```python
nums, k, numOperations = [1,90], 76, 1

nums.sort()
n = len(nums)
l = 0
ans = 1

if n == 2:
    if nums[1] - nums[0] < k:
        return 2
    else:
        return 1
for r in range(n):
    while nums[r] - nums[l] > 2 * k:
        l += 1
    window_size = r - l + 1
    if window_size > numOperations + 1:
        current = numOperations + 1
    else:
        current = window_size

    ans = max(ans, current)


print(ans)
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_nums, input_k, input_numOperations, expected):
    result = solution.maxFrequency(input_nums, input_k, input_numOperations)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - (nums, k, numOperations): {input_nums, input_k, input_numOperations}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [1,4,5], 1, 2, 2),
    (2, [5,11,20,20], 5, 1, 2),
]

for case_number, input_nums, input_k, input_numOperations, expected in test_cases:
    run_test_case(case_number, input_nums, input_k, input_numOperations, expected)
```
