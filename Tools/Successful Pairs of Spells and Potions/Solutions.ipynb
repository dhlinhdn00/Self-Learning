# **SOLUTION**

```python
class Solution(object):
    def successfulPairs(self, spells, potions, success):
        """
        :type spells: List[int]
        :type potions: List[int]
        :type success: int
        :rtype: List[int]
        """
        return self.solverBinarySearch(spells, potions, success)

    def solverNaive(self, spells, potions, success):
        """
        :cost: TLE
        """
        n = len(spells)
        ans = [0] * n
        m = len(potions)
        potions.sort()
        for i in range(n):

            if potions[m - 1] * spells[i] < success:
                continue
            if potions[0] * spells[i] >= success:
                ans[i] = m
            else:
                reversed_count = 1
                j = 1
                while j < m and potions[j] * spells[i] < success:
                    reversed_count += 1
                    j += 1
                ans[i] = (m - reversed_count)
        return ans

    def solverTwoPointers(self, spells, potions, success):
        """
        :cost: 301 ms (82.46%) | 34.03 MB (6.82%)
        """
        n = len(spells)
        m = len(potions)
        potions.sort()
        ans = [0] * n

        indexed_spells = sorted([(spells[i], i) for i in range(n)], key=lambda x: x[0])

        j = m - 1
        for spell, idx in indexed_spells:
            while j >= 0 and potions[j] * spell >= success:
                j -= 1
            ans[idx] = m - (j + 1)

        return ans

    def solverBinarySearch(self, spells, potions, success):
        """
        :cost: 240 ms (93.57%) | 37.44 MB (14.03%)
        """
        import bisect
        n = len(spells)
        m = len(potions)
        potions.sort()
        ans = [0] * n

        for i in range(n):
            spell = spells[i]
            min_potion = (success + spell - 1) // spell 
            idx = bisect.bisect_left(potions, min_potion)
            ans[i] = m - idx

        return ans
```

```python
solution = Solution()

def run_test_case(case_number, input_spells, input_potions, input_success, expected):
    result = solution.successfulPairs(input_spells, input_potions, input_success)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - (spells, potions, sucess): ({input_spells, input_potions, input_success}), Output: {result}, Expected: {expected}, Status: {status}")
test_cases = [
    (1, [5,1,3], [1,2,3,4,5], 7, [4,0,3]),
    (2, [3,1,2], [8,5,8], 16, [2,0,2])
]

for case_number, input_spells, input_potions, input_success, expected in test_cases:
    run_test_case(case_number, input_spells, input_potions, input_success, expected)
```
