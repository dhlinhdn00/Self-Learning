# **SOLUTION**

```python
import heapq

class Solution(object):
    def processQueries(self, c, connections, queries):
        """
        :type c: int
        :type connections: List[List[int]]
        :type queries: List[List[int]]
        :rtype: List[int]
        """

        parent = list(range(c + 1))
        size = [1] * (c + 1)
        online = [True] * (c + 1)
        heaps = [None] + [[i] for i in range(1, c + 1)]  

        def find(x):
            if parent[x] != x:
                parent[x] = find(parent[x])
            return parent[x]

        def union(a, b):
            pa, pb = find(a), find(b)
            if pa == pb:
                return
            if size[pa] < size[pb]:
                pa, pb = pb, pa
            parent[pb] = pa
            size[pa] += size[pb]
            if len(heaps[pb]) > len(heaps[pa]):
                heaps[pa], heaps[pb] = heaps[pb], heaps[pa]
            for v in heaps[pb]:
                heapq.heappush(heaps[pa], v)
            heaps[pb] = []

        for u, v in connections:
            union(u, v)

        res = []
        for t, x in queries:
            if t == 1:
                p = find(x)
                while heaps[p] and not online[heaps[p][0]]:
                    heapq.heappop(heaps[p])
                if heaps[p]:
                    res.append(heaps[p][0])
                else:
                    res.append(-1)
            else:
                online[x] = False  

        return res
```

```python
solution = Solution()

# def extraCondition(result, expected):
#     pass
def run_test_case(case_number, input_c, input_connections, input_queries,
 expected):
    result = solution.processQueries(input_c, input_connections, input_queries)
    status = "PASSED" if result == expected else "FAILED"
    # status = "PASSED" if extraCondition(result, expected) else "FAILED"

    print(f"Case {case_number} - (c, connections, queries): {input_c, input_connections, input_queries}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, 5, [[1,2],[2,3],[3,4],[4,5]], [[1,3],[2,1],[1,1],[2,2],[1,2]], [3,2,3]),
    (2, 3, [], [[1,1],[2,1],[1,1]], [1,-1]),
]

for case_number, input_c, input_connections, input_queries, expected in test_cases:
    run_test_case(case_number, input_c, input_connections, input_queries, expected)
```
