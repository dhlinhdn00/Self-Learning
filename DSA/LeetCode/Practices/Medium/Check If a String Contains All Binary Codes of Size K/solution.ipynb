# **SOLUTION**

```python
class Solution(object):
    def hasAllCodes(self, s, k):
        """
        :type s: str
        :type k: int
        :rtype: bool
        """
        return self.solverSlidingWusingBuiltin(s, k)

    def solverSlidingW(self, s, k):
        """
        :cost: 154ms (100%) | 30.40MB (96.36%)
        """
        n = len(s)

        if n < k:
            return False
        
        # Step 1: Total numbers of different binary codes of lenght k
        need = 1 << k # same as 2**k but faster

        # Step 2: Quick Impossibility check. Number turn for sliding window is (n - k + 1)
        if n - k + 1 < need:
            return False
        
        # Step 3: Mask keeps only the last k bits of rolling value.
        mask = need - 1
        count = 0

        # Step 4: Seen array
        seen = [False] * need

        # Step 5: Build the integer value for the first window of size k
        value = 0
        for i in range(k):
            # Shift left (multiply by 2) and add current bit (0 or 1) -> just the convert binary number to decimal
            value = (value << 1) | (s[i] == "1")

        # Step 6: Mark the first code as seen.
        if not seen[value]:
            seen[value] = True
            count += 1
            if count == need:
                return True

        # Step 7: Slide window
        for i in range(k, n):
            bit = 1 if s[i] == "1" else 0
            value = ((value << 1) & mask | bit) # & mask helps us keep all and remove leftmost (older bit), nice!
        
        # Step 8: Check
            if not seen[value]:
                seen[value] = True
                count += 1
                if count == need:
                    return True
        

        return False

    def solverSlidingWusingBuiltin(self, s, k):
        """
        :cost: 309ms (41.82%) | 30.22MB (98.18%)
        """
        n = len(s)
        if n < k:
            return False
        
        need = 1 << k
        if n - k + 1 < need:
            return False
        
        seen = [False] * need
        count = 0
        
        for i in range(n - k + 1):
            value = int(s[i:i+k], 2)
            if not seen[value]:
                seen[value] = True
                count += 1
                if count == need:
                    return True

        return False


        
```
