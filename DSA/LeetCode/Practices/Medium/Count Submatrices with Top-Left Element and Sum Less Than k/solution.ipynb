# **SOLUTION**

```python
class Solution(object):
    def countSubmatrices(self, grid, k):
        """
        :type grid: List[List[int]]
        :type k: int
        :rtype: int
        """
        return self.solverPrefixSum(grid, k)

    def solverPrefixSum(self, grid, k):
        h, w = len(grid), len(grid[0])

        count = 0
        prefixSum = [[0] * w for _ in range(h)]
        # prefixSum[0][0] = grid[0][0]

        for i in range(h):
            for j in range(w):
                if i == 0 and j == 0:
                    prefixSum[i][j] = grid[0][0]
                    if grid[0][0] == 0:
                        count += 1
                elif i == 0 and j > 0:
                    prefixSum[i][j] = grid[i][j] + prefixSum[i][j - 1]
                elif i > 0:
                    if prefixSum[i - 1][j] == 0:
                        prefixSum[i][j] = 0
                    else:
                        prefixSum[i][j] = grid[i][j] + prefixSum[i - 1][j] + prefixSum[i][j - 1]
                        if j != 0:
                            prefixSum[i][j] -= prefixSum[i - 1][j - 1]

                if prefixSum[i][j] > 0:
                    count += 1
                if prefixSum[i][j] > k:
                    prefixSum[i][j] = 0
                    count -= 1
                    break
        return count
        
```
