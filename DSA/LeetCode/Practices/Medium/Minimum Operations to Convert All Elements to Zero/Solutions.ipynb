# **SOLUTION**

```python
class Solution(object):
    def minOperations(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        return self.solverStack(nums)

    def solverStack(self, nums):       
        ans = 0
        stack = [0]  

        for x in nums:
            while stack and stack[-1] > x:
                stack.pop()
            if stack[-1] < x:
                ans += 1
                stack.append(x)

        return ans
```

```python
solution = Solution()

# def extraCondition(result, expected):
#     pass

def run_test_case(case_number, input_nums, expected):
    result = solution.minOperations(input_nums)
    status = "PASSED" if result == expected else "FAILED"
    # status = "PASSED" if extraCondition(result, expected) else "FAILED"

    print(f"Case {case_number} - nums: {input_nums}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [0, 2], 1),
    (2, [3,1,2,1], 3),
    (3, [1,2,1,2,1,2], 4)
]

for case_number, input_nums, expected in test_cases:
    run_test_case(case_number, input_nums, expected)
```
