# **SOLUTION**

```python
class Solution(object):
    def minimumTotal(self, triangle):
        """
        :type triangle: List[List[int]]
        :rtype: int
        """
        return self.solverDP(triangle)
    
    def solverDP(self, triangle):
        n = len(triangle)
        dp = triangle[-1][:]

        for i in range(n - 2, -1, -1):
            for j in range(len(triangle[i])):
                dp[j] = triangle[i][j] + min(dp[j], dp[j + 1])
        
        return dp[0]
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_triangle, expected):
    result = solution.minimumTotal(input_triangle)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - triangle: {input_triangle}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [[2],[3,4],[6,5,7],[4,1,8,3]], 11),
    (2, [[-10]], -10),
]

for case_number, input_triangle, expected in test_cases:
    run_test_case(case_number, input_triangle, expected)
```
