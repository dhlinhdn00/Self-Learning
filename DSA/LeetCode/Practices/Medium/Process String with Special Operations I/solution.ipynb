# **SOLUTION**

```python
class Solution(object):
    def processStr(self, s):
        """
        :type s: str
        :rtype: str
        """
        return self.solverRuleBased(s)
    
    def solverRuleBased(self, s):
        ans = ""
        
        for word in s:
            if ord(word) >= 97 and ord(word) <= 122:
                ans += word
            elif word == "*":
                if ans:
                    ans = ans[:-1]
            elif word == "%":
                ans = ans[::-1]
            elif word == "#":
                ans += ans

        return ans
```
