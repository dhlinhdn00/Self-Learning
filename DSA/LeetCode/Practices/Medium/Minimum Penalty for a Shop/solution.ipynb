# **SOLUTION**

```python
class Solution(object):
    def bestClosingTime(self, customers):
        """
        :type customers: str
        :rtype: int
        """
        return self.solverPrefixSumOptimized(customers)

    def solverPrefixSum(self, customers):
        ":cost: 107ms (59.52%) | 19.25MB (34.52%)"
        n = len(customers)
        nP = customers.count("Y")
        p = [nP] * (n + 1)

        for i in range(1, n + 1):
            if customers[i - 1] == "Y":
                p[i] = p[i - 1] - 1
            else:
                p[i] = p[i - 1] + 1
        return p.index(min(p))

    def solverPrefixSumOptimized(self, customers):
        ":cost: 70 ms (76.18%) | 16.59MB (47.62%)"
        n = len(customers)
        p = customers.count("Y")
        min_p = p
        min_idx = 0
        for i in range(n):
            if customers[i] == "Y":
                p -= 1
            else:
                p += 1

            if p < min_p:
                min_p = p
                min_idx = i + 1
        return min_idx
```
