# **SOLUTION**

```python
class Solution(object):
    def maximumAmount(self, coins):
        """
        :type coins: List[List[int]]
        :rtype: int
        """
        return self.solverDP(coins)
    
    def solverDP(self, coins):
        h, w = len(coins), len(coins[0])
        NEG = float('-inf')

        dp = [[[NEG, NEG, NEG] for _ in range(w)] for _ in range(h)]

        if coins[0][0] >= 0:
            dp[0][0][0] = coins[0][0]
            dp[0][0][1] = 0

        for i in range(h):
            for j in range(w):
                if i == 0 and j == 0:
                    continue

                val = coins[i][j]
                for k in range(3):
                    if i > 0 and dp[i - 1][j][k] != NEG:
                        dp[i][j][k] = max(dp[i][j][k], dp[i - 1][j][k] + val)
                    if j > 0 and dp[i][j - 1][k] != NEG:
                        dp[i][j][k] = max(dp[i][j][k], dp[i][j - 1][k] + val)

                    if val < 0 and k > 0:
                        if i > 0 and dp[i - 1][j][k - 1] != NEG:
                            dp[i][j][k] = max(dp[i][j][k], dp[i - 1][j][k - 1])
                        if j > 0 and dp[i][j - 1][k - 1] != NEG:
                            dp[i][j][k] = max(dp[i][j][k], dp[i][j - 1][k - 1])
        
        return max(dp[h - 1][w - 1])

        
```
