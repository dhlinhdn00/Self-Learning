# **SOLUTION**

```python
class Solution(object):
    def minTime(self, skill, mana):
        """
        :type skill: List[int]
        :type mana: List[int]
        :rtype: int
        """
        return self.solverDPOptimized(skill, mana)

    def solverDP(self, skill, mana):
        """
        :cost: TLE
        """
        n = len(skill)
        dp = [0] * n
        dp[0] = skill[0] * mana[0]
        for i in range(1, n):
            dp[i] = dp[i - 1] + skill[i] * mana[0]

        for potion in mana[1:]:
            temp = dp[:]
            temp[0] = dp[1]
            i = 0
            while i < n:
                if temp[i] >= dp[i + 1]:
                    temp[i + 1] = temp[i] + skill[i] * potion
                    i += 1
                else:
                    gap = dp[i + 1] - temp[i]
                    temp[:i + 1] = [x + gap for x in temp[:i + 1]]
            dp = temp[:]

        return dp[-1]
    
    def solverDPOptimized(self, skill, mana):

        n = len(skill)

        A = [0] * (n + 1)
        for i in range(n):
            A[i + 1] = A[i] + skill[i]

        dp = [0] * n
        for i in range(n):
            dp[i] = A[i + 1] * mana[0]

        for potion in mana[1:]:
            S = dp[0] - A[0] * potion 
            for i in range(1, n):
                cand = dp[i] - A[i] * potion
                if cand > S:
                    S = cand

            temp = [0] * n
            for i in range(n):
                temp[i] = S + A[i + 1] * potion

            dp = temp  

        return dp[-1]
    
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_skill, input_mana, expected):
    result = solution.minTime(input_skill, input_mana)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - (skill, mana): ({input_skill, input_mana}), Output: {result}, Expected: {expected}, Status: {status}")
test_cases = [
    (1, [1,5,2,4], [5,1,4,2], 110),
    (2, [1,1,1], [1,1,1], 5),
    (3, [1,2,3,4], [1,2], 21),
]

for case_number, input_skill, input_mana, expected in test_cases:
    run_test_case(case_number, input_skill, input_mana, expected)
```
