# **SOLUTION**

```python
class Solution(object):
    def maxIceCream(self, costs, coins):
        """
        :type costs: List[int]
        :type coins: int
        :rtype: int
        """
        return self.solverNaive(costs, coins)
    def solverNaive(self, costs, coins):
        n = len(costs)
        costs.sort()

        for i in range(n):
            coins -= costs[i]
            if coins < 0:
                return i
        return n
        
```
