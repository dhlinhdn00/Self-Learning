# **SOLUTION**

```python
class Solution(object):
    def reorderedPowerOf2(self, n):
        """
        :type n: int
        :rtype: bool
        """
        constraint = 10**9
        return self.solverNaive(n, constraint)
    
    def solverNaive(self, n, constraint):
        powers = [1 << e for e in range((constraint - 1).bit_length())]
        sorted_chars = [''.join(sorted(str(x))) for x in powers]
        n_str = ''.join(sorted(str(n)))
        if n_str in sorted_chars:
            return True
        return False
        
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_n, expected):
    result = solution.reorderedPowerOf2(input_n)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - n: {input_n}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, 1, True),
    (2, 10, False),
]

for case_number, input_n, expected in test_cases:
    run_test_case(case_number, input_n, expected)
```
