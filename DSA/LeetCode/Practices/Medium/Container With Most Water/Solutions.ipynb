# **SOLUTION**

```python
class Solution(object):
    def maxArea(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        return self.solver2Pointers(height)
    
    def solver2Pointers(self, height):
        l, r = 0, len(height) - 1
        max_area = 0
        while l < r:
            hl, hr = height[l], height[r]
            max_area = max (max_area, (r -l) * min(hr, hl))
            if hr < hl:
                r -= 1
            else:
                l += 1
        return max_area
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_height, expected):
    result = solution.maxArea(input_height)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - height: {input_height}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [1,8,6,2,5,4,8,3,7], 49),
    (2, [1, 1], 1),
]

for case_number, input_height, expected in test_cases:
    run_test_case(case_number, input_height, expected)
```
