# **SOLUTION**

```python
class Solution(object):
    def numberOfSubstrings(self, s):
        """
        :type s: str
        :rtype: int
        """
        return self.solverSlidingWindow(s)
    
    def solverSlidingWindow(self, s):
        count = {'a' : 0, 'b' : 0, 'c' : 0}
        n = len(s)
        l = 0
        ans = 0

        for r in range(n):
            count[s[r]] += 1

            while count['a'] * count['b'] * count['c'] > 0:
                ans += n - r
                count[s[l]] -= 1
                l += 1

        return ans
        
```
