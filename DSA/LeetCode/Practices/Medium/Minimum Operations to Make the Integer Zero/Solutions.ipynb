# **SOLUTION**

```python
class Solution(object):
    def makeTheIntegerZero(self, num1, num2):
        """
        :type num1: int
        :type num2: int
        :rtype: int
        """
        return self.solverNaive(num1, num2)

    def solverNaive(self, num1, num2):
        # num1 = 2^i + num2 with k times => k(num1 - num2) = sigma(2^i)
        for k in range(1, 61):
            S = num1 - k * num2
            if S < 0:
                continue
            if S >= k and bin(S).count("1") <= k:
                return k
            
        return -1
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_num1, input_num2, expected):
    result = solution.makeTheIntegerZero(input_num1, input_num2)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - (num1, num2): {input_num1, input_num2}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, 3, -2, 3),
    (2, 5, 7, -1),
]

for case_number, input_num1, input_num2, expected in test_cases:
    run_test_case(case_number, input_num1, input_num2, expected)
```
