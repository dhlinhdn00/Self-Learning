# **SOLUTION**

```python
class Solution(object):
    def xorAfterQueries(self, nums, queries):
        """
        :type nums: List[int]
        :type queries: List[List[int]]
        :rtype: int
        """
        return self.solverBF(nums, queries, 1000000007)

    def solverBF(self, nums, queries, *args):
        for l, r, k, v in queries:
            i = l
            while i <= r:
                nums[i] = (nums[i] * v) % args[0]
                i += k

        # ans = 0
        # for num in nums:
        #     ans ^= num

        ans = nums[0]
        for num in nums[1:]:
            ans ^= num

        return ans
```
