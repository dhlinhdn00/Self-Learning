# **SOLUTION**

```python
class Solution(object):
    def countMajoritySubarrays(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        return self.solverPrefixSum(nums, target)
    
    def solverPrefixSum(self, nums, target):
        n = len(nums)

        mapped_nums = [1 if num == target else -1 for num in nums]

        prefixSum = [0] * (n + 1)
        for i in range(n):
            prefixSum[i + 1] = prefixSum[i] + mapped_nums[i]

        ans = 0
        for i in range(1, n + 1):
            for j in range(i, n + 1):
                if prefixSum[j] > prefixSum[i - 1]:
                    ans += 1
        return ans
```
