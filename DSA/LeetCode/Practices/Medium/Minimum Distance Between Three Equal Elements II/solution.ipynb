# **SOLUTION**

```python
class Solution(object):
    def minimumDistance(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        return self.solverNaive(nums)

    def solverNaive(self, nums):
        positions = {}

        for idx, val in enumerate(nums):
            if val not in positions:
                positions[val] = []
            positions[val].append(idx)

        best_span = float('inf')

        for val in positions:
            pos = positions[val]
            count = len(pos)

            if count < 3:
                continue

            for i in range(count - 2):
                span = pos[i + 2] - pos[i]
                best_span = min(best_span, span)

        if best_span == float('inf'):
            return -1

        return best_span * 2
```
