# **SOLUTION**

```python
class Solution(object):
    def countCollisions(self, directions):
        """
        :type directions: str
        :rtype: int
        """
        return self.solverMath(directions)

    def solverGreedyState(self, directions):
        """
        :cost: 259ms (46.15%) | 13.17MB (46.15%)
        """
        prev_dir = directions[0]
        count = 0
        potential_colision_R = 0

        for dir in directions[1:]:
            if prev_dir == "R" and dir == "L":
                count += (2 + potential_colision_R)
                prev_dir = "S"
                potential_colision_R = 0
            elif prev_dir == "S" and dir == "L":
                count += 1
                prev_dir = "S"
            elif prev_dir == "R" and dir == "S":
                count += (1 + potential_colision_R)
                prev_dir = "S"
                potential_colision_R = 0
            elif prev_dir == "R" and dir == "R":
                potential_colision_R += 1
                prev_dir = "R"
            else:
                prev_dir = dir

        return count
    
    def solverMath(self, directions):
        """
        :cost: 120ms (73.08%) | 13.26MB (42.31%)
        """
        n = len(directions)

        i = 0
        while i < n and directions[i] == "L":
            i += 1
        
        j = n - 1
        while j >= 0 and directions[j] == "R":
            j -= 1

        return sum(c != "S" for c in directions[i:j+1])
```
