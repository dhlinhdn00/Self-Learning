# **SOLUTION**

```python
class Solution(object):
    def sumFourDivisors(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        return self.solverMath(nums)

    def primeLessThan(self, x):
        sieve = [True] * x
        sieve[0] = sieve[1] = False
        for p in range(2, int(x**0.5) + 1):
            if sieve[p]:
                for m in range(p * p, x, p):
                    sieve[m] = False
        return sieve
    
    def solverMath(self, nums):
        MAX = 10**5 + 1
        is_prime = self.primeLessThan(MAX)

        ans = 0
        for n in nums:
            case_1_check = False
            # case 1: n = p^3
            p = int(round(n ** (1.0/3.0)))
            for cand in (p - 1, p, p + 1):
                if cand > 1 and cand * cand * cand == n and is_prime[cand]:
                    ans += 1 + cand + cand * cand + n
                    case_1_check = True
                    break
            if case_1_check:
                continue
            # case 2: n = p * q (p != q)
            limit = int(n**0.5)
            for p in range(2, limit + 1):
                if n % p == 0:
                    q = n // p
                    if q != p and is_prime[p] and is_prime[q]:
                        ans += 1 + p + q + n
                        break
        return ans
```
