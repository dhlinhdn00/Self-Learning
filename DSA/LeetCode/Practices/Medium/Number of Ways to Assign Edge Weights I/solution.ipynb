# **SOLUTION**

```python
from collections import deque
class Solution(object):
    def assignEdgeWeights(self, edges):
        """
        :type edges: List[List[int]]
        :rtype: int
        """
        MOD = 10**9 + 7
        return self.solverBFSMath(edges, MOD)
    
    def solverBFSMath(self, edges, MOD):
        """
        Math Trick: 
            In Fact, this problem just reduces to finding the maximum depth of the tree, 
            then we select the weight of edge from root to any node at maximum depth.
            for example, we find x edges to reach the max depth, from 1 to x - 1, we have 2 options,
            with x, we have just 1 option to ensure the total cost is odd.
            So, the result is 2^(x-1)
        """
        nE = len(edges)
        nV = nE + 1

        adjList = {i : [] for i in range(1, nV + 1)}

        for u, v in edges:
            adjList[u].append(v)
            adjList[v].append(u)
        
        max_depth = 0

        queue =deque([(1, 0)]) # node, depth
        visited = {1}

        while queue:
            node, depth = queue.popleft()
            
            for neighbor in adjList[node]:
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append((neighbor, depth + 1))

            
            max_depth = max(depth, max_depth)

        if max_depth == 0:
            return 0
        return pow(2, max_depth - 1, MOD)


        
```
