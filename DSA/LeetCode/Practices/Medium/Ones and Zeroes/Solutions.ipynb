# **SOLUTION**

```python
class Solution(object):
    def findMaxForm(self, strs, m, n):
        """
        :type strs: List[str]
        :type m: int
        :type n: int
        :rtype: int
        """
        return self.solverDP(strs, m, n)
    
    def solverGreedy(self, strs, m, n):
        """
        :status: Failed
        """
        data = [[s.count("0"), s.count("1")] for s in strs]

        if m > n:
            sorted_data = sorted(data, key=lambda x: (x[1], x[0] + x[1]))
        else:
            sorted_data = sorted(data, key=lambda x: (x[0], x[0] + x[1]))
        
        check_m, check_n, count = 0, 0, 0
        for d in sorted_data:
            if check_m + d[0] <= m and check_n + d[1] <= n:
                check_m += d[0]
                check_n += d[1]
                count += 1
            else:
                continue
        return count 
    
    def solverDP(self, strs, m, n):
        dp = [[0] * (n + 1) for _ in range(m + 1)]

        for s in strs:
            zeros = s.count("0")
            ones = s.count("1")
            for i in range(m, zeros - 1, -1):
                for j in range(n, ones - 1, -1):
                    dp[i][j] = max(dp[i][j], dp[i - zeros][j - ones] + 1)

        return dp[m][n]
```
