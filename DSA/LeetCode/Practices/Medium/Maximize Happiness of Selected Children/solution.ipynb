# **SOLUTION**

```python
class Solution(object):
    def maximumHappinessSum(self, happiness, k):
        """
        :type happiness: List[int]
        :type k: int
        :rtype: int
        """
        return self.solverNaive(happiness, k)

    def solverNaive(self, happiness, k):
        happiness.sort(reverse=True)

        ord_happiness = [max(0, happiness[i] - i) for i in range(len(happiness))]

        return sum(ord_happiness[:k])
        
```
