# **SOLUTION**

```python
class Solution(object):
    def maxSubarraySum(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: int
        """
        return self.solverPrefixSum(nums, k)

    def solverPrefixSum(self, nums, k):
        n = len(nums)
        pref = [0] * (n + 1)
        hashmap = {}
        hashmap.setdefault(0, []).append(0)
        
        for i in range(1, n + 1):
            pref[i] = pref[i - 1] + nums[i - 1]
            r = i % k
            if r not in hashmap:
                hashmap[r] = []
            hashmap[r].append(i)
        
        maxsum = float("-inf")
        for indices in hashmap.values():
            if len(indices) >= 2:
                min_pref = pref[indices[0]]
                for idx in indices[1:]:
                    cur = pref[idx] - min_pref
                    if cur > maxsum:
                        maxsum = cur
                    if pref[idx] < min_pref:
                        min_pref = pref[idx]

        return maxsum
        
```

```python
# Make Prefix Sum easier! Basic Prefix Sum: Using a query for Prefix Sum

def rangeSumQueries(nums, queries):
    n = len(nums)
    prefix_sum = [0] * (n + 1)

    for i in range(1, n + 1):
        prefix_sum[i] = prefix_sum[i - 1] + nums[i - 1]

    result = []

    for l, r in queries:
        total = prefix_sum[r + 1] - prefix_sum[l]
        result.append(total)

    return result


nums = [2, -1, 3, 5]
queries = [(0, 1), (1, 3), (0, 3)]

print(rangeSumQueries(nums, queries))
```
