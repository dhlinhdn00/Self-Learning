# **SOLUTION**

```python
class Solution(object):
    def numberOfSubmatrices(self, grid):
        """
        :type grid: List[List[str]]
        :rtype: int
        """
        return self.solverPrefix(grid)

    def solverPrefix(self, grid):
        h, w = len(grid), len(grid[0])
        prefixGrid =[[0] * w for _ in range(h)]
        prefixNonZero = [[0] * w for _ in range(h)]
        count = 0
        for i in range(h):
            for j in range(w):
                if grid[i][j] == "X":
                    val = 1
                elif grid[i][j] == "Y":
                    val = -1
                else:
                    val = 0

                prefixGrid[i][j] = val
                if i > 0:
                    prefixGrid[i][j] += prefixGrid[i - 1][j]
                if j > 0:
                    prefixGrid[i][j] += prefixGrid[i][j - 1]
                if i > 0 and j > 0:
                    prefixGrid[i][j] -= prefixGrid[i - 1][j - 1]

                nonZero = 1 if val != 0 else 0
                prefixNonZero[i][j] = nonZero
                if i > 0:
                    prefixNonZero[i][j] += prefixNonZero[i - 1][j]
                if j > 0:
                    prefixNonZero[i][j] += prefixNonZero[i][j - 1]
                if i > 0 and j > 0:
                    prefixNonZero[i][j] -= prefixNonZero[i - 1][j - 1]

                if prefixGrid[i][j] == 0 and prefixNonZero[i][j] > 0:
                    count += 1
        return count
```
