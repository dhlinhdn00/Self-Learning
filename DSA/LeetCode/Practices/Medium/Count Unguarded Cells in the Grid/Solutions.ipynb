# **SOLUTION**

```python
class Solution(object):
    def countUnguarded(self, m, n, guards, walls):
        """
        :type m: int
        :type n: int
        :type guards: List[List[int]]
        :type walls: List[List[int]]
        :rtype: int
        """
        return self.solverNaive(m, n, guards, walls)
    
    def solverNaive(self, m, n, guards, walls):
        grid = [[0] * n for i in range(m)]
        for i, j in guards:
            grid[i][j] = 1
        for i, j in walls:
            grid[i][j] = 2

        def update(cell, guards_check):
            if cell == 1:
                return 1, True
            elif cell == 2:
                return 2, False
            else:
                if guards_check:
                    if cell != -1:
                        return -1, True
                    else:
                        return cell, True
                else:
                    return cell, False

        for i in range(m):
            guards_check = False
            for j in range(n):
                cell = grid[i][j]
                grid[i][j], guards_check = update(cell, guards_check)

        for i in range(m):
            guards_check = False
            for j in range(n - 1, -1, -1):
                cell = grid[i][j]
                grid[i][j], guards_check = update(cell, guards_check)

        for j in range(n):
            guards_check = False
            for i in range(m):
                cell = grid[i][j]
                grid[i][j], guards_check = update(cell, guards_check)

        for j in range(n):
            guards_check = False
            for i in range(m - 1, -1, -1):
                cell = grid[i][j]
                grid[i][j], guards_check = update(cell, guards_check)

        return sum(row.count(0) for row in grid)
```
