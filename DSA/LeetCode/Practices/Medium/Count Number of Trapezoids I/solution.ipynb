# **SOLUTION**

```python
class Solution(object):
    def countTrapezoids(self, points):
        """
        :type points: List[List[int]]
        :rtype: int
        """
        MODULO = 10**9 + 7
        return self.solverMath(points, MODULO)

    def solverMath(self, points, MODULO):
        ycount = {}
        for x, y in points:
            if y in ycount:
                ycount[y] += 1
            else:
                ycount[y] = 1

        def combination2(n):
            return ((n * (n - 1)) // 2) % MODULO

        combs = []
        for value in ycount.values():
            combs.append(combination2(value))

        # ans = 0
        # nc = len(comb_list)
        # for i in range(nc - 1):
        #     for j in range(i + 1, nc):
        #         ans = (ans + comb_list[i] * comb_list[j]) % MODULO

        # -----------------------------------------------------------
        # Explanation of why we avoid the O(k^2) double loop:
        #
        # We want the sum of all pairwise products:
        #     T = Σ_{i<j} (C_i * C_j)
        #
        # A naive double loop is O(k^2).
        #
        # Let:
        #     S  = C1 + C2 + ... + Ck
        #     S2 = C1^2 + C2^2 + ... + Ck^2
        #
        # Now expand S^2:
        #     S^2 = (C1 + C2 + ... + Ck)^2
        #         = C1^2 + C2^2 + ... + Ck^2
        #           + 2*(C1C2 + C1C3 + ... all i<j pairs)
        #
        # The term we need, T, is exactly inside that expansion:
        #     2*T = S^2 - S2
        #       T = (S^2 - S2) / 2
        #
        # -----------------------------------------------------------

        S = sum(combs) % MODULO
        S2 = sum((c * c) % MODULO for c in combs) % MODULO

        ans = (S * S - S2) % MODULO
        ans = ans * pow(2, MODULO - 2, MODULO) % MODULO   # division by 2 via modular inverse

        return ans
        
```
