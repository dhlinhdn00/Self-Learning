# **SOLUTION**

```python
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def createLinkedListFromList(arr):
    if not arr:
        return None
    head = ListNode(arr[0])
    curr = head
    
    for val in arr[1:]:
        curr.next = ListNode(val)
        curr = curr.next

    return head

def createListFromLinkedList(head):
    result = []
    curr = head
    while curr:
        result.append(curr.val)
        curr = curr.next
    return result

class Solution(object):
    def deleteMiddle(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        head = createLinkedListFromList(head)
        ans = self.solverSlowFast(head)
        
        return createListFromLinkedList(ans)

    def solverSlowFast(self, head):
        if not head.next:
            return None
        prev = ListNode(None)
        slow, fast = head, head

        while fast and fast.next:
            prev = slow
            slow = slow.next
            fast = fast.next.next


            # print("After Iter:")
            # print(f"slow = {slow.val}, fast = {fast.val}, prev = {prev.val}")

        prev.next = slow.next
        return head
```

# **DRAFT**

```python
input_head = [1,3,4,7,1,2,6]
solution = Solution()

result = solution.deleteMiddle(input_head)
print(result)
```
