# **SOLUTION**

```python
class Solution(object):
    def getDescentPeriods(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """
        return self.solverDP(prices)
    
    def solverDP(self, prices):
        n = len(prices)
        dp = [0] * n
        dp[0] = 1

        for i in range(n):
            if prices[i - 1] - prices[i] == 1:
                dp[i] = dp[i - 1] + 1
            else:
                dp[i] = 1
        return sum(dp)
```
