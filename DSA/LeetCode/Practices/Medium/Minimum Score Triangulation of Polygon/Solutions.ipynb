# **SOLUTION**

```python
class Solution(object):
    def minScoreTriangulation(self, values):
        """
        :type values: List[int]
        :rtype: int
        """
        return self.solverDP(values)

    def solverDP(self, values):
        n = len(values)
        dp = [[0] * n for _ in range(n)]
        for length in range(3, n + 1):
            for i in range(n - length + 1):
                j = i + length - 1
                dp[i][j] = float("inf")
                for k in range(i + 1, j):
                    score = dp[i][k] + dp[k][j] + values[i] * values[k] * values[j]
                    dp[i][j] = min(dp[i][j], score)
                 
        return dp[0][n - 1]
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_values, expected):
    result = solution.minScoreTriangulation(input_values)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - (values): {input_values}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [1,2,3], 6),
    (2, [3,7,4,5], 144),
    (3, [1,3,1,4,1,5], 13),
]

for case_number, input_values, expected in test_cases:
    run_test_case(case_number, input_values, expected)
```
