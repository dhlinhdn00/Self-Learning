# **SOLUTION**

```python
class Solution(object):
    def largestSubmatrix(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: int
        """
        return self.solverGreedy(matrix)
    
    def solverGreedy(self, matrix):
        h, w = len(matrix), len(matrix[0])
        prefixMatrix = [row[:] for row in matrix]
        maxArea = 0
        
        for i in range(1, h):
            for j in range(w):
                if prefixMatrix[i][j] == 1:
                    prefixMatrix[i][j] += prefixMatrix[i - 1][j]
                else:
                    prefixMatrix[i][j] = 0
        
        for row in prefixMatrix:
            row.sort(reverse=True)
            for i, v in enumerate(row):
                maxArea = max(maxArea, v * (i + 1))
        return maxArea
```
