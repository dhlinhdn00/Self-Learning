# **SOLUTION**

```python
class Solution(object):
    def angleClock(self, hour, minutes):
        """
        :type hour: int
        :type minutes: int
        :rtype: float
        """
        return self.solverMath(hour, minutes)

    def solverMath(self, hour, minutes):
        if hour == 12:
            hour = 0
        hourMinutes = hour + minutes / 60.0
        hourMinutesPos = hourMinutes * 30
        minutesPos = minutes * 6

        angle = round(abs(minutesPos - hourMinutesPos), 5)
        
        return angle if angle <= 180 else 360 - angle
```
