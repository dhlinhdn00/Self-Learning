# **SOLUTION**

```python
import numpy as np

class Solution(object):
    def maxMatrixSum(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: int
        """
        return self.solverGreedyNumpy(matrix)

    def solverGreedy(self, matrix):
        ":cost: 152ms(10.53%) | 18.44MB(63.16%)"
        h = len(matrix)
        w = len(matrix[0])
        is_positive = 1
        min_abs = float("inf")
        sum_abs = 0
        for i in range(h):
            for j in range(w):
                num = matrix[i][j]
                min_abs = min(min_abs, abs(num))
                sum_abs += abs(num)
                if num == 0:
                    is_positive *= 1
                else:
                    is_positive *= (num / abs(num))
        if is_positive == 1:
            return sum_abs
        return sum_abs - 2 * min_abs

    def solverGreedyNumpy(self, matrix):
        ":cost: 23ms(100%) | 30.79MB(5.26%)"
        arr = np.array(matrix)
        abs_arr = np.abs(arr)
        sum_abs = abs_arr.sum()
        min_abs = abs_arr.min()

        neg_count = np.sum(arr < 0)

        if neg_count % 2 == 0:
            return sum_abs
        return sum_abs - 2 * min_abs
```
