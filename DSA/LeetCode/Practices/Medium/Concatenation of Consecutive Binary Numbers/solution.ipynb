# **SOLUTION**

```python
class Solution(object):
    def concatenatedBinary(self, n):
        """
        :type n: int
        :rtype: int
        """
        MOD = 10**9 + 7
        return self.solverNaive(n, MOD)
    
    def solverNaive(self, n, MOD):
        ans = 0
        for i in range(1, n + 1):
            bit_len = i.bit_length()
            ans = ((ans << bit_len) + i ) % MOD
        return ans
```
