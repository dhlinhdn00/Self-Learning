# **SOLUTION**

```python
class Solution(object):
    def maxSumDivThree(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        return self.solverDP(nums)
    
    def solverGreedy(self, nums):
        """
        :cost: 26ms (76%) | 15.14MB (32%)
        """
        dic = {1 : [], 2 : []}
        ans = 0
        for num in nums:
            ans += num
            if num % 3 != 0:
                dic[num % 3].append(num)
        if ans % 3 == 1:
            dic[1].sort(), dic[2].sort()
            ans -= min(dic[1][0] if len(dic[1
            ]) >= 1 else float("inf"), sum(dic[2][:2]) if len(dic[2]) >= 2 else float("inf"))
        elif ans % 3 == 2:
            dic[1].sort(), dic[2].sort()
            ans -= min(dic[2][0] if len(dic[2]) >= 1 else float("inf"), sum(dic[1][:2]) if len(dic[1]) >= 2 else float("inf"))
        return ans
    
    def solverDP(self, nums):
        """
        :cost: 185ms (16%) | 15.12MB (32%)
        """
        dp = [0, float("-inf"), float("-inf")]
        for num in nums:
            ndp = dp[:]
            for r in range(3):
                if dp[r] != float("-inf"):
                    v = dp[r] + num
                    nr = v % 3
                    ndp[nr] = max(ndp[nr], v)
            dp = ndp

        return dp[0]

```
