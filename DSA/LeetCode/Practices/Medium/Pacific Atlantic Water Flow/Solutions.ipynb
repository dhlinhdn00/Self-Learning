# **SOLUTION**

```python
class Solution(object):
    def pacificAtlantic(self, heights):
        """
        :type heights: List[List[int]]
        :rtype: List[List[int]]
        """
        return self.solverDP(heights)

    def solverDP(self, heights):
        h = len(heights)
        w = len(heights[0])
        ans = []

        # Pacific: top-left
        dpPacific = []
        for i in range(h):
            row = [0] * w
            if i == 0:
                row = [1] * w
            else:
                row[0] = 1
            dpPacific.append(row)
        
        for i in range(1, h):
            for j in range(1, w):
                if ((heights[i][j] >= heights[i][j - 1] and dpPacific[i][j - 1] == 1) or
                    (heights[i][j] >= heights[i - 1][j] and dpPacific[i - 1][j] == 1)):
                    dpPacific[i][j] = 1

        self._stabilize(dpPacific, heights)

        # Atlantic: bottom-right
        dpAtlantic = []
        for i in range(h):
            row = [0] * w
            if i == h - 1:
                row = [1] * w
            else:
                row[w - 1] = 1
            dpAtlantic.append(row)

        for i in range(h - 2, -1, -1):
            for j in range(w - 2, -1, -1):
                if ((heights[i][j] >= heights[i][j + 1] and dpAtlantic[i][j + 1] == 1) or
                    (heights[i][j] >= heights[i + 1][j] and dpAtlantic[i + 1][j] == 1)):
                    dpAtlantic[i][j] = 1

        self._stabilize(dpAtlantic, heights)

        for i in range(h):
            for j in range(w):
                if dpAtlantic[i][j] == 1 and dpPacific[i][j] == 1:
                    ans.append([i, j])
        return ans

    def _stabilize(self, dp, heights):
        h = len(heights)
        w = len(heights[0])
        changed = True
        while changed:
            changed = False
            for i in range(h):
                for j in range(w):
                    if dp[i][j] == 1:
                        continue
                    if ((i > 0 and dp[i - 1][j] == 1 and heights[i][j] >= heights[i - 1][j]) or
                        (j > 0 and dp[i][j - 1] == 1 and heights[i][j] >= heights[i][j - 1]) or
                        (i + 1 < h and dp[i + 1][j] == 1 and heights[i][j] >= heights[i + 1][j]) or
                        (j + 1 < w and dp[i][j + 1] == 1 and heights[i][j] >= heights[i][j + 1])):
                        dp[i][j] = 1
                        changed = True
```
