# **SOLUTION**

```python
class Solution(object):
    def minOperations(self, nums):

        return self.solverNaive(nums)

    def solverNaive(self, nums):
        def gcd(n1, n2):
            while n2:
                n1, n2 = n2, n1 % n2
            return n1

        n = len(nums)

        # check gcd for all    
        g = nums[0]
        for x in nums:
            g = gcd(g, x)
        if g > 1:
            return -1

        # if there are "1" in nums
        ones = nums.count(1)
        if ones > 0:
            return n - ones

        # there is no "1" in nums, find the min len which has gcd = 1
        best = float('inf')
        for i in range(n):
            g = nums[i]
            for j in range(i + 1, n):
                g = gcd(g, nums[j])
                if g == 1:
                    best = min(best, j - i + 1)
                    break

        return (best - 1) + (n - 1)
```
