# **SOLUTION**

```python
class Solution(object):
    def minSubarray(self, nums, p):
        """
        :type nums: List[int]
        :type p: int
        :rtype: int
        """
        return self.solverPrefixSum(nums, p)
    
    def solverPrefixSum(self, nums, p):
        total = sum(nums)
        need = total % p
        if need == 0:
            return 0 

        pos = {0: -1}       
        curr = 0
        res = len(nums)

        for i, x in enumerate(nums):
            curr = (curr + x) % p
            target = (curr - need) % p

            if target in pos:
                res = min(res, i - pos[target])  

            pos[curr] = i 

        return res if res < len(nums) else -1

```

```python
nums, p = [3,1,4,2], 6

n = len(nums)
pref = [0] * (n + 1)

for i in range(1, n + 1):
    pref[i] = pref[i - 1] + nums[i - 1]
pref
```

```python
solution = Solution()

# def extraCondition(result, expected):
#     pass


def run_test_case(case_number, input_nums, input_p, expected):
    result = solution.minSubarray(input_nums, input_p)
    status = "PASSED" if result == expected else "FAILED"
    # status = "PASSED" if extraCondition(result, expected) else "FAILED"
    print(f"Case {case_number} - (nums, p): {(input_nums, input_p)}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [3,1,4,2], 6, 1),
    (2, [6,3,5,2], 9, 2),
    (3, [1,2,3], 3, 0),
]

for case_number, input_nums, input_p, expected in test_cases:
    run_test_case(case_number, input_nums, input_p, expected)
```
