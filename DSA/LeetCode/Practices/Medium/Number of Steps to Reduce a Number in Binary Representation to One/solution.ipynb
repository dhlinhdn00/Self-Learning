# **SOLUTION**

```python
class Solution(object):
    def numSteps(self, s):
        """
        :type s: str
        :rtype: int
        """
        return self.solverNaive(s)
    
    def solverNaive(self, s):
        n = int(s, 2)
        plus1 = 0
        div2 = 0
        total = 0
        while n !=1:
            if n % 2 == 1:
                n += 1
                plus1 += 1
                total +=1
            n //= 2
            div2 += 1
            total += 1

        return total
```

```python
solution = Solution()

# def extraCondition(result, expected):
#     pass


def run_test_case(case_number, input_s, expected):
    result = solution.numSteps(input_s)
    status = "PASSED" if result == expected else "FAILED"
    # status = "PASSED" if extraCondition(result, expected) else "FAILED"
    print(f"Case {case_number} - (s): {(input_s)}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, "1101", 6),
    (2, "10", 1),
    (3, "1", 0),
]

for case_number, input_s, expected in test_cases:
    run_test_case(case_number, input_s, expected)
```
