# **SOLUTION**

```python
class Solution(object):
    def maximumElementAfterDecrementingAndRearranging(self, arr):
        """
        :type arr: List[int]
        :rtype: int
        """
        return self.solverGreedy(arr)
    
    def solverGreedy(self, arr):
        ans = 0
        arr.sort()

        for a in arr:
            if ans < a:
                ans += 1

        return ans
```
