# **SOLUTION**

```python
class Solution(object):
    def countPermutations(self, complexity):
        """
        :type complexity: List[int]
        :rtype: int
        """
        return self.solverMath(complexity)
    
    def solverMath(self, complexity):
        MOD = 10**9 + 7
        def factorial(n):
            result = 1
            for i in range(2, n + 1):
                result = (result * i) % MOD
            return result

        n = len(complexity)
        root = complexity[0]
        for c in complexity[1:]:
            if c <= root:
                return 0

        return factorial(n - 1)


        
```
