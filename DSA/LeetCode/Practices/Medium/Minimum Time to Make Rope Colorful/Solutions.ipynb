# **SOLUTION**

```python
class Solution(object):
    def minCost(self, colors, neededTime):
        """
        :type colors: str
        :type neededTime: List[int]
        :rtype: int
        """
        return self.solverNaive(colors, neededTime)

    def solverNaive(self, colors, neededTime):
        
        n = len(colors)
        start = 0
        count = 1
        ans = 0

        for i in range(1, n):
            if colors[i] == colors[i - 1]:
                count += 1
            else:
                if count >= 2:
                    ans += sum(neededTime[start : i]) - max(neededTime[start : i])

                start = i
                count = 1
        if count >= 2:
            ans += sum(neededTime[start : n]) - max(neededTime[start : n])

        return ans
```

```python
solution = Solution()

# def extraCondition(result, expected):
#     pass

def run_test_case(case_number, input_colors, input_neededTime, expected):
    result = solution.minCost(input_colors, input_neededTime)
    status = "PASSED" if result == expected else "FAILED"
    # status = "PASSED" if extraCondition(result, expected) else "FAILED"

    print(f"Case {case_number} - (colors, neededTime): {input_colors, input_neededTime}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, "abaac", [1,2,3,4,5], 3),
    (2, "abc", [1,2,3], 0),
    (3, "aabaa", [1,2,3,4,1], 2)
]

for case_number, input_colors, input_neededTime, expected in test_cases:
    run_test_case(case_number, input_colors, input_neededTime, expected)
```
