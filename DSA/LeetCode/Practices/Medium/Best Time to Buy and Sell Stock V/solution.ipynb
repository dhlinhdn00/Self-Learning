# **SOLUTION**

```python
class Solution(object):
    def maximumProfit(self, prices, k):
        return self.solverDP(prices, k)

    def solverDP(self, prices, k):
        n = len(prices)
        NEG = -float('inf')

        dp = [[NEG] * 3 for _ in range(k + 1)]
        dp[0][0] = 0
        dp[0][1] = -prices[0]
        dp[0][2] =  prices[0]

        for i in range(1, n):
            price = prices[i]
            prev = [row[:] for row in dp] 

            for t in range(k + 1):

                dp[t][0] = prev[t][0]
                if t > 0:
                    dp[t][0] = max(dp[t][0],
                                   prev[t-1][1] + price,   # sell long
                                   prev[t-1][2] - price)   # buy back short

                dp[t][1] = max(prev[t][1],
                               prev[t][0] - price)

                dp[t][2] = max(prev[t][2],
                               prev[t][0] + price)

        return max(dp[t][0] for t in range(k + 1))
```
