# **SOLUTION**

```python
class Solution(object):
    def minimumDistance(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        return self.solverNaive(nums)
    
    def solverNaive(self, nums):
        freq = {}
        first_pos = {}
        last_pos = {}

        for idx, num in enumerate(nums):
            if num in freq:
                freq[num] += 1
                last_pos[num] = idx
            else:
                freq[num] = 1
                first_pos[num] = idx
                last_pos[num] = idx

        max_count = 0
        max_value = None
        max_span = -1

        for k in freq:
            count = freq[k]
            span = last_pos[k] - first_pos[k]

            if count > max_count:
                max_count = count
                max_span = span
                max_value = k
            elif count == max_count:
                if span < max_span:
                    max_span = span
                    max_value = k
        if max_count < 3:
            return -1
        return max_span * 2
```

# **DRAFT**

```python
nums = [1,1,2,3,2,1,2]

freq = {}
first_pos = {}
last_pos = {}

for idx, num in enumerate(nums):
    if num in freq:
        freq[num] += 1
        last_pos[num] = idx
    else:
        freq[num] = 1
        first_pos[num] = idx
        last_pos[num] = idx

max_count = 0
max_value = None
max_span = -1

for k in freq:
    count = freq[k]
    span = last_pos[k] - first_pos[k]

    if count > max_count:
        max_count = count
        max_span = span
        max_value = k
    elif count == max_count:
        if span < max_span:
            max_span = span
            max_value = k

print(max_span * 2)
```
