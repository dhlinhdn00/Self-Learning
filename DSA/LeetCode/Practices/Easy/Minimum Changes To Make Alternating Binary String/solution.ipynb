# **SOLUTION**

```python
class Solution(object):
    def minOperations(self, s):
        """
        :type s: str
        :rtype: int
        """
        return self.solverOptimized(s)
    def solverNaive(self, s):
        """
        :cost: 15ms(72.04%) | 12.58MB(77.42%)
        """
        if len(s) % 2 == 0:
            n = len(s)
        else:
            n = len(s) - 1
        count01 = sum(s[i:i+2] == "01" for i in range(0, n, 2))
        count10 = sum(s[i:i+2] == "10" for i in range(0, n, 2))
        ans = n // 2 - count01 - count10
        if count01 > count10:
            ans += count10 * 2
            fin = "1"
        else:
            ans += count01 * 2
            fin= "0"
        
        if n != len(s) and s[-1] == fin and count01 != count10:
            return ans + 1
        return ans
    def solverOptimized(self, s):
        """
        :cost: 12ms(78.49%) | 12.50MB(95.70%)
        """
        diff0 = 0
        diff1 = 0
        for i, ch in enumerate(s):
            # 01 ... 
            expect0 = "0" if i % 2 == 0 else "1"
            if ch != expect0:
                diff0 += 1
            # 10 ...
            else:
                diff1 += 1
        return min(diff0, diff1)
    
```
