# **SOLUTION**

```python
class Solution(object):
    def maxNumberOfBalloons(self, text):
        """
        :type text: str
        :rtype: int
        """
        return self.solverNaive(text)
    
    def solverNaive(self, text):
        """
        cost: 4ms(68.52%) | 12.28MB(98.61%)
        """
        rootDict = {}
        textDict = {}
        comp = []

        for c in "balloon":
            rootDict[c] = rootDict.get(c, 0) + 1
            textDict[c] = 0

        for c in text:
            if c in "balloon":
                textDict[c] += 1

        for c in rootDict.keys():
            comp.append(textDict[c] // rootDict[c])

        return min(comp)
```
