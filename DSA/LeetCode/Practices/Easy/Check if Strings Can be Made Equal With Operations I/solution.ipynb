# **SOLUTION**

```python
class Solution(object):
    def canBeEqual(self, s1, s2):
        """
        :type s1: str
        :type s2: str
        :rtype: bool
        """
        return self.solverNaive(s1, s2)

    def solverNaive(self, s1, s2):
        if set([s1[0], s1[2]]) != set([s2[0], s2[2]]):
            return False
        if set([s1[1], s1[3]]) != set([s2[1], s2[3]]):
            return False
        return True
```

# **TESTCASES**

```python
solution = Solution()

# def extraCondition(result, expected):
#     pass


def run_test_case(case_number, input_s1, input_s2, expected):
    result = solution.canBeEqual(input_s1, input_s2)
    status = "PASSED" if result == expected else "FAILED"
    # status = "PASSED" if extraCondition(result, expected) else "FAILED"
    print(f"Case {case_number} - (s1, s2): {(input_s1, input_s2)}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, "abcd", "cdab", True),
    (2, "abcd", "dacb", False),
]

for case_number, input_s1, input_s2, expected in test_cases:
    run_test_case(case_number, input_s1, input_s2, expected)
```
