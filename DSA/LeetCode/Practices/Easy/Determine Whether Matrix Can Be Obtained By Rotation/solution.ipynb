# **SOLUTION**

```python
class Solution(object):
    def findRotation(self, mat, target):
        """
        :type mat: List[List[int]]
        :type target: List[List[int]]
        :rtype: bool
        """
        return self.solverNaive(mat, target)
    
    def solverNaive(self, mat, target):
        """
        :cost: 3ms (35.92%) | 12.38MB (77.67%)"""
        n = len(mat)
        check0, check90, check180, check270 = True, True, True, True
        for i in range(n):
            for j in range(n):
                if mat[i][j] != target[i][j]: check0 = False
                if mat[n - j - 1][i] != target[i][j]: check90 = False
                if mat[n - i - 1][n - j - 1] != target[i][j]: check180 = False
                if mat[j][n - i - 1] != target[i][j]: check270 = False

        return (check0 or check90 or check180 or check270)
```
