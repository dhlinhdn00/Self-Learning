# **SOLUTION**

```python
class Solution(object):
    def reverseSubmatrix(self, grid, x, y, k):
        """
        :type grid: List[List[int]]
        :type x: int
        :type y: int
        :type k: int
        :rtype: List[List[int]]
        """
        return self.solverNaive(grid, x, y, k)

    def solverNaive(self, grid, x, y, k):
        """
        :cost: 1ms(48.15%) | 12.55MB (88.89%)
        """
        neo = 2 * x + k - 1

        temp = [0] * y
        for i in range(x, x + k // 2):
            temp = grid[i][y : y + k]
            grid[i][y : y + k] = grid[neo - i][y : y + k]
            grid[neo - i][y : y + k] = temp
        return grid
        
```
