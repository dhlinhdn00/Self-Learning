# **SOLUTION**

```python
import numpy as np

class Solution(object):
    def countNegatives(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        return self.solverNumpy(grid)

    def solverNaive(self, grid):
        """:cost: 14ms (6.10%) | 24.45MB (1.81%)"""
        m, n = len(grid), len(grid[0])
        
        count = 0
        for i in range(m):
            for j in range(n):
                if grid[i][j] < 0:
                    count += 1
        return count


    def solverNumpy(self, grid):
        """:cost: 0ms (100%) | 24.63MB (1.81%)"""
        grid_np = np.array(grid)
        return (np.array(grid) < 0).sum()


```
