# **SOLUTION**

```python
class Solution(object):
    def mapWordWeights(self, words, weights):
        """
        :type words: List[str]
        :type weights: List[int]
        :rtype: str
        """
        return self.solverNaive(words, weights)

    def solverNaive(self, words, weights):
        """
        :notice: ord("a") = 97; ord("z") = 122
        """
        ans = ""
        for word in words:
            sumWord = 0
            for char in word:
                sumWord += weights[ord(char) - 97]
            ans += chr(122 - (sumWord % 26))
        return ans
                
        
```
