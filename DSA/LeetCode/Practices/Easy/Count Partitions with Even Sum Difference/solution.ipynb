# **SOLUTION**

```python
class Solution(object):
    def countPartitions(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        return self.solverNaive(nums)

    def solverNaive(self, nums):
        if sum(nums) % 2 == 0:
            return len(nums) - 1
        return 0
```
