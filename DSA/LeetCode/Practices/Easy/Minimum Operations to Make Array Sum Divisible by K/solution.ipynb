# **SOLUTION**

```python
class Solution(object):
    def minOperations(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: int
        """
        return self.solverNaive(nums, k)
    
    def solverNaive(self, nums, k):
        return sum(nums) % k
        
```

```python
solution = Solution()

# def extraCondition(result, expected):
#     pass


def run_test_case(case_number, input_nums, input_k, expected):
    result = solution.minOperations(input_nums, input_k)
    status = "PASSED" if result == expected else "FAILED"
    # status = "PASSED" if extraCondition(result, expected) else "FAILED"
    print(f"Case {case_number} - (nums, k): {(input_nums, input_k)}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [3,9,7], 5, 4),
    (2, [4,1,3], 4, 0),
    (3, [3,2], 6, 5),
]

for case_number, input_nums, input_k, expected in test_cases:
    run_test_case(case_number, input_nums, input_k, expected)
```
