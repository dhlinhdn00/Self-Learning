# **SOLUTION**

```python
class Solution(object):
    def minimumCost(self, cost):
        """
        :type cost: List[int]
        :rtype: int
        """
        return self.solverNaive(cost)
    
    def solverNaive(self, cost):
        cost.sort(reverse=True)
        total = 0

        for i, v in enumerate(cost):
            if i % 3 != 2:
                total += v
        return total
        
```

```python
solution = Solution()

# def extraCondition(result, expected):
#     pass


def run_test_case(case_number, input_cost, expected):
    result = solution.minimumCost(input_cost)
    status = "PASSED" if result == expected else "FAILED"
    # status = "PASSED" if extraCondition(result, expected) else "FAILED"
    print(f"Case {case_number} - (cost): {(input_cost)}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [1,2,3], 5),
    (2, [6,5,7,9,2,2], 23),
    (3, [5,5], 10),
]

for case_number, input_cost, expected in test_cases:
    run_test_case(case_number, input_cost, expected)
```
