# **SOLUTION**

```python
class Solution(object):
    def minimumOperations(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        return self.solverNaive(nums)
    def solverNaive(self, nums):
        ans = 0
        for num in nums:
            if num % 3 != 0:
                ans += 1
        return ans
```
