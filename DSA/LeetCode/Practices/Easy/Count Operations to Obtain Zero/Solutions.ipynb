# **SOLUTION**

```python
class Solution(object):
    def countOperations(self, num1, num2):
        """
        :type num1: int
        :type num2: int
        :rtype: int
        """
        return self.solverNaive(num1, num2)

    def solverNaive(self, num1, num2):
        min_n, max_n = min(num1, num2), max(num1, num2)
        count = 0

        while min_n > 0:
            n = max_n // min_n
            max_n, min_n = min_n, max_n % min_n
            count += n

        return count
```

```python
solution = Solution()

# def extraCondition(result, expected):
#     pass

def run_test_case(case_number, input_num1, input_num2, expected):
    result = solution.countOperations(input_num1, input_num2)
    status = "PASSED" if result == expected else "FAILED"
    # status = "PASSED" if extraCondition(result, expected) else "FAILED"

    print(f"Case {case_number} - (input1, input2): {input_num1, input_num2}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, 2, 3, 3),
    (2, 10, 10, 1),
]

for case_number, input_num1, input_num2, expected in test_cases:
    run_test_case(case_number, input_num1, input_num2, expected)
```
