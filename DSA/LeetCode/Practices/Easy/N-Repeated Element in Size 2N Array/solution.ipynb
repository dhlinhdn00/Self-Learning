# **SOLUTION**

```python
class Solution(object):
    def repeatedNTimes(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        return self.solverNaive(nums)

    def solverNaive(self, nums):
        n = len(nums)
        target = n // 2
        
        dic = {}

        for i in range(n):
            if nums[i] in dic:
                dic[nums[i]] += 1
            else:
                dic[nums[i]] = 1
        
        for key, value in dic.items():
            if value == target:
                return key
```
