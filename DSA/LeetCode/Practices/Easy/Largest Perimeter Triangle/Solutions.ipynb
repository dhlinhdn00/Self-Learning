# **SOLUTION**

```python
class Solution(object):
    def largestPerimeter(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        return self.solverNaive(nums)
    
    def solverNaive(self, nums):
        nums.sort(reverse=True)
        n = len(nums)
        for i in range(n - 2):
            a = nums[i]
            b = nums[i + 1]
            c = nums[i + 2]
            if a < b + c:
                print(a, b, c)
                return a + b + c
        return 0
```

```python
solution = Solution()

def run_test_case(case_number, input_nums, expected):
    result = solution.largestPerimeter(input_nums)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - fruits: {input_nums}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [2,1,2], 5),
    (2, [1,2,1,10], 0),
]

for case_number, input_nums, expected in test_cases:
    run_test_case(case_number, input_nums, expected)
```
