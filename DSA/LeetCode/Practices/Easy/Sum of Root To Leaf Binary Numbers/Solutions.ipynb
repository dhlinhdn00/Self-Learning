# **SOLUTION**

```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


class Solution(object):
    def sumRootToLeaf(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: int
        """
        def dfs(node, current):
            if not node:
                return 0
            
            current = current * 2 + node.val

            if not node.left and not node.right:
                return current

            return dfs(node.left, current) + dfs(node.right, current)

        return dfs(root, 0)
```

# **TESTCASES**

```python
from collections import deque

def build_tree(values):
    if not values:
        return None
    
    root = TreeNode(values[0])
    queue = deque([root])
    i = 1
    
    while queue and i < len(values):
        node = queue.popleft()
        
        if i < len(values) and values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1
        
        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1
    
    return root


solution = Solution()

def run_test_case(case_number, input_root, expected):
    tree_root = build_tree(input_root)
    result = solution.sumRootToLeaf(tree_root)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - Input: {input_root}, Output: {result}, Expected: {expected}, Status: {status}")


test_cases = [
    (1, [1,0,1,0,1,0,1], 22),
    (2, [0], 0),
    (3, [1,1], 3), 
]

for case_number, input_root, expected in test_cases:
    run_test_case(case_number, input_root, expected)
```
