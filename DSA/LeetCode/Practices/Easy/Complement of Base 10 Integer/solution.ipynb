# **SOLUTION**

```python
class Solution(object):
    def bitwiseComplement(self, n):
        """
        :type n: int
        :rtype: int
        """
        return self.solverMath(n)

    def solverNaive(self, n):
        """
        cost: 0ms(100%) | 12.40MB (82.14%)
        """
        binary = bin(n)[2:]
        flip_binary = ''.join('1' if bit == '0' else '0' for bit in binary)
        return int(flip_binary, 2)
        
    def solverMath(self, n):
        """
        cost: 0ms(100%) | 12.40MB (82.14%)
        """
        if n == 0:
            return 1
        return ((1 << n.bit_length()) - 1) ^ n
        
```
