# **SOLUTION**

```python
class Solution(object):
    def numOfStrings(self, patterns, word):
        """
        :type patterns: List[str]
        :type word: str
        :rtype: int
        """
        return self.solverNaive(patterns, word)

    def solverNaive(self, patterns, word):
        ans = 0
        for p in patterns:
            if p in word:
                ans += 1
        return ans
```

# **TESTCASES**

```python
solution = Solution()

# def extraCondition(result, expected):
#     pass


def run_test_case(case_number, input_patterns, input_word, expected):
    result = solution.numOfStrings(input_patterns, input_word)
    status = "PASSED" if result == expected else "FAILED"
    # status = "PASSED" if extraCondition(result, expected) else "FAILED"
    print(f"Case {case_number} - (patterns, word): {(input_patterns, input_word)}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, ["a","abc","bc","d"], "abc", 3),
    (2, ["a","b","c"], "aaaaabbbbb", 2),
    (3, ["a","a","a"], "ab", 3),
]

for case_number, input_patterns, input_word, expected in test_cases:
    run_test_case(case_number, input_patterns, input_word, expected)
```
