# **SOLUTION**

```python
class Solution(object):
    def minimumBoxes(self, apple, capacity):
        """
        :type apple: List[int]
        :type capacity: List[int]
        :rtype: int
        """
        return self.solverNaive(apple, capacity)

    def solverNaive(self, apple, capacity):
        capacity.sort(reverse=True)
        s = sum(apple)
        check_cap = 0
        n = len(capacity)
        
        for i in range(n):
            check_cap += capacity[i]

            if check_cap >= s:
                return i + 1
                
        return n
```
