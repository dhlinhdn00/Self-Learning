# **SOLUTION**

```python
class Solution(object):
    def numSpecial(self, mat):
        """
        :type mat: List[List[int]]
        :rtype: int
        """
        return self.solverNaive(mat)
    def solverNaive(self, mat):
        h, w = len(mat), len(mat[0])
        row = [0] * h
        col = [0] * w
        for i in range(h):
            for j in range(w):
                if mat[i][j] == 1:
                    row[i] += 1
                    col[j] += 1
        ans = 0
        for i in range(h):
            for j in range(w):
                if mat[i][j] == 1:
                    if row[i] == 1 and col[j] == 1:
                        ans += 1
        return ans
                    


```
