# **SOLUTION**

```python
class Solution(object):
    def getSneakyNumbers(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        return self.solverNaive(nums)
    
    def solverNaive(self, nums):
        dic = {}

        for num in nums:
            if num not in dic:
                dic[num] = 1
            else:
                dic[num] += 1

        ans = [k for k, v in dic.items() if v == 2]

        return ans
```

# **TESTCASES**

```python
solution = Solution()

def extraCondition(result, expected):
    return set(result).issubset(set(expected))

def run_test_case(case_number, input_n, expected):
    result = solution.getSneakyNumbers(input_n)
    status = "PASSED" if extraCondition(result, expected) else "FAILED"
    print(f"Case {case_number} - n: {input_n}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [0,1,1,0], [0, 1]),
    (2, [0,3,2,1,3,2], [2, 3]),
    (3, [7,1,5,4,3,4,6,0,9,5,8,2], [4, 5])
]

for case_number, input_n, expected in test_cases:
    run_test_case(case_number, input_n, expected)
```
