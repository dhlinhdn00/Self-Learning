# **SOLUTION**

```python
class Solution:
    def earliestFinishTime(self, landStartTime, landDuration, waterStartTime, waterDuration):
        return self.solverNaive(landStartTime, landDuration, waterStartTime, waterDuration)
    def solverNaive(self, landStartTime, landDuration, waterStartTime, waterDuration):
        landL = len(landStartTime)
        waterL = len(waterStartTime)

        earliestTime = float("inf")

        for i in range(landL):
            for j in range(waterL):
                # case1: land first
                landEnd = landStartTime[i] + landDuration[i]
                waterBegin = max(landEnd, waterStartTime[j])
                finish1 = waterBegin + waterDuration[j]

                # case2: water first
                waterEnd = waterStartTime[j] + waterDuration[j]
                landBegin = max(waterEnd, landStartTime[i])
                finish2 = landBegin + landDuration[i]

                earliestTime = min(earliestTime, finish1, finish2)

        return earliestTime
```

# **DRAFT**

```python
landStartTime, landDuration = [2,8], [4,1]
waterStartTime, waterDuration = [3], [3]
"""
landStartTime = 3; landDuration = 2 => 5
waterStartTime = 
"""

landL = len(landStartTime)
waterL = len(waterStartTime)
earliestTime = float("inf")
for i in range(landL):
    for j in range(waterL):
        landTime = landStartTime[i] + landDuration[i]
        waterTime = waterStartTime[j] + waterDuration[j]

        if landTime < waterStartTime[j]:
            totalTime = waterTime
        else:
            totalTime = landTime

    earliestTime = min(totalTime, earliestTime)

print(earliestTime)
```
