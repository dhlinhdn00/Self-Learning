# **SOLUTION**

```python
import re

class Solution(object):
    def validateCoupons(self, code, businessLine, isActive):
        """
        :type code: List[str]
        :type businessLine: List[str]
        :type isActive: List[bool]
        :rtype: List[str]
        """
        return self.solverNaive(code, businessLine, isActive)
    
    def solverNaive(self, code, businessLine, isActive):
        pattern = r"^[A-Za-z0-9_]+$"

        valid_lines = ["electronics", "grocery", "pharmacy", "restaurant"]
        order = {line: i for i, line in enumerate(valid_lines)}

        valid = []

        for c, b, active in zip(code, businessLine, isActive):

            # 3. isActive[i] is true.

            if not active:
                continue

            # 1. code[i] is non-empty and consists only of alphanumeric characters (a-z, A-Z, 0-9) and underscores (_).
            # Why? Because in constraint, 0 <= code[i].length, businessLine[i].length <= 100

            if not c:
                continue

            if not re.match(pattern, c):
                continue
            
            # 2. businessLine[i] is one of the following four categories: "electronics", "grocery", "pharmacy", "restaurant".
            if b not in order:
                continue

            valid.append((order[b], c))

        valid.sort()
        return [c for _, c in valid]
        
```
