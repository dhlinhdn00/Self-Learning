# **SOLUTION**

```python
class Solution(object):
    def largestAltitude(self, gain):
        """
        :type gain: List[int]
        :rtype: int
        """
        return self.solverPrefix(gain)

    def solverPrefix(self, gain):
        """
        cost: 0ms (100%) | 12.20MB (99.44%)
        """
        prefix = [0]
        prev = prefix[0]
        for g in gain:
            prefix.append(prev + g)
            prev += g
        return max(prefix)
```
