# **SOLUTION**

```python
class Solution(object):
    def countTriples(self, n):
        """
        :type n: int
        :rtype: int
        """
        return self.solverLoop(n)
    
    def solverLoop(self, n):
        powers = []
        for i in range(1, n + 1):
            powers.append(i ** 2)

        ans = 0
        for i in range(n - 1):
            for j in range(i, n):
                c = powers[i] + powers[j]
                if c in powers:
                    if i != j:
                        ans += 2
                    else:
                        ans += 1

        return ans
```
