# **SOLUTION**

```python
class Solution(object):
    def countOdds(self, low, high):
        """
        :type low: int
        :type high: int
        :rtype: int
        """
        return self.solverMath(low, high)


    def solverMath(self, low, high):
        return (high + 1)//2 - low//2
```

# **TESTCASES**

```python
solution = Solution()

# def extraCondition(result, expected):
#     pass


def run_test_case(case_number, input_low, input_high, expected):
    result = solution.countOdds(input_low, input_high)
    status = "PASSED" if result == expected else "FAILED"
    # status = "PASSED" if extraCondition(result, expected) else "FAILED"
    print(f"Case {case_number} - (low, high): {(input_low, input_high)}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, 3, 7, 3),
    (2, 8, 10, 1),
]

for case_number, input_low, input_high, expected in test_cases:
    run_test_case(case_number, input_low, input_high, expected)
```
