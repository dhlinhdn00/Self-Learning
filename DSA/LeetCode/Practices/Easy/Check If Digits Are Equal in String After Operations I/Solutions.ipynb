# **SOLUTION**

```python
class Solution(object):
    def hasSameDigits(self, s):
        """
        :type s: str
        :rtype: bool
        """
        return self.solverNaive(s)

    def solverNaive(self, s):
        n = len(s)
        nums = [int(ch) for ch in s]
        check_num = n
        while check_num != 3:
            check_num -=1
            for i in range(check_num):
                nums[i] = (nums[i] + nums[i + 1]) % 10
        return True if nums[0] == nums[2] else False
        
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_s, expected):
    result = solution.hasSameDigits(input_s)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - s: {input_s}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, "3902", True),
    (2, "34789", False,
]

for case_number, input1, input2, expected in test_cases:
    run_test_case(case_number, input1, input2, expected)
```
