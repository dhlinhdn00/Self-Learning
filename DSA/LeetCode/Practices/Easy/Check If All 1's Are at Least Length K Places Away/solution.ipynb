# **SOLUTION**

```python
class Solution(object):
    def kLengthApart(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: bool
        """

        return self.solverNaive(nums, k)
    
    def solverNaive(self, nums, k):
        check = False
        count = 0
        for num in nums:
            if num == 0 and check == False:
                continue
            elif num == 1 and check == False:
                check = True
            elif num == 0 and check == True:
                count += 1
            elif num == 1 and check == True:
                if count < k:
                    return False
                else:
                    count = 0
                    

        return True
                
```
