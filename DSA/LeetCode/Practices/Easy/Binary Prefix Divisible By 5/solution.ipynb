# **SOLUTION**

```python
class Solution(object):
    def prefixesDivBy5(self, nums):
        """
        :type nums: List[int]
        :rtype: List[bool]
        """
        return self.solverLoop(nums)
    
    def solverLoop(self, nums):
        n = len(nums)
        accum = 0
        ans = []
        for num in nums:
            accum += num * (2 ** n)
            ans.append(accum % 5 == 0)
            n -= 1
        return ans
        
```

# **DRAFT**

```python
nums = [0,1,1]
n = len(nums)
accum = 0
ans = []
for num in nums:
    accum += num * 2**n
    ans.append(accum % 5 == 0)
    n -= 1
ans
```
