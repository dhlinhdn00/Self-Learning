# **SOLUTION**

```python
class Solution(object):
    def minimumDifference(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: int
        """
        return self.solverSlidingWindow(nums, k)
    
    def solverSlidingWindow(self, nums, k):
        if k == 1:
            return 0
        n = len(nums)
        nums.sort()

        min_distance = float("inf")
        for i in range(n - k + 1):
            min_distance = min(min_distance, nums[i + k - 1] - nums[i])

        return min_distance
        
```
