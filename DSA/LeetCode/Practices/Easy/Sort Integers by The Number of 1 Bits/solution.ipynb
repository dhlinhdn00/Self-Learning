# **SOLUTION**

```python
class Solution(object):
    def sortByBits(self, arr):
        """
        :type arr: List[int]
        :rtype: List[int]
        """
        return self.solverFast(arr)
    
    def solverNaive(self, arr):
        """
        cost: 1023ms (5.55%) | 12.63MB (38.33%)
        """
        arr.sort()
        n = len(arr)
        # buble sort
        for i in range(n):
            for j in range(0, n - i - 1):
                if bin(arr[j]).count("1") > bin(arr[j + 1]).count("1"):
                    arr[j], arr[j + 1] = arr[j + 1], arr[j]

        return arr
    
    def solverFast(self, arr):
        """
        cost: 4ms (93.33%) | 12.73MB (6.11%)
        """
        return sorted(arr, key=lambda x: (bin(x).count("1"), x))
```
