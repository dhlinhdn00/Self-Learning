# **SOLUTION**

```python
class Solution(object):
    def processStr(self, s, k):
        """
        :type s: str
        :type k: int
        :rtype: str
        """
        return self.solverRuleBased(s, k)

    def solverRuleBased(self, s, k):
        length = 0

        for word in s:
            if word == "*":
                length = max(0, length - 1)

            elif word == "%":
                pass

            elif word == "#":
                length *= 2

            else:
                length += 1

        if k >= length:
            return "."

        for i in range(len(s) - 1, -1, -1):
            word = s[i]

            if word == "*":

                length += 1

            elif word == "%":

                k = length - 1 - k

            elif word == "#":
                old_length = length // 2


                if k >= old_length:
                    k -= old_length

                length = old_length

            else:
                length -= 1

                if k == length:
                    return word

        return "."
```

# **DRAFT**

```python
s = '%#bz%xum##i##vzo#pwc*#dkwbh####%uf%s*%cgppqhqa%h#l##o%ij%%cz%iga##e###u%#e####jfwx##%%*x%m*%#'
k = 6572

result = ""
count = 0

for word in s:
    if ord(word) >= 97 and ord(word) <= 122:
        result += word
        count += 1
    elif word == "*":
        if result:
            result = result[:-1]
            count -= 1
    elif word == "%":
        result = result[::-1]
    elif word == "#":
        result += result
        count *= 2
print(count)
if k <= count - 1:
    print(result[k])

print(".")
```

# **TESTCASES**

```python
solution = Solution()

# def extraCondition(result, expected):
#     pass


def run_test_case(case_number, input_s, input_k, expected):
    result = solution.processStr(input_s, input_k)
    status = "PASSED" if result == expected else "FAILED"
    # status = "PASSED" if extraCondition(result, expected) else "FAILED"
    print(f"Case {case_number} - (s, k): {(input_s, input_k)}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, "a#b%*", 1, "a"),
    (2, "cd%#*#", 3, "d"),
    (3, "z*#", 0, "."),
]

for case_number, input_s, input_k, expected in test_cases:
    run_test_case(case_number, input_s, input_k, expected)
```
