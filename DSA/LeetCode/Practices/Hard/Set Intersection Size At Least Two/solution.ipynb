# **SOLUTION**

```python
class Solution(object):
    def intersectionSizeTwo(self, intervals):
        """
        :type intervals: List[List[int]]
        :rtype: int
        """
        return self.solverGreedy(intervals)

    def solverGreedy(self, intervals):
        intervals.sort(key=lambda x: (x[1], -x[0]))
        
        a = b = -10**18  
        res = 0    
        
        for l, r in intervals:
            if l > b:
                res += 2
                a = r - 1
                b = r
            elif l > a:
                res += 1
                a = b
                b = r
            else:
                pass
        
        return res
```
