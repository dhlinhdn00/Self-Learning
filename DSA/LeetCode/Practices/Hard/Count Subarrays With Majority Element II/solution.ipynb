# **SOLUTION**

```python
class Solution(object):
    def countMajoritySubarrays(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        return self.solverPrefixSumOptimized(nums, target)
    
    def solverPrefixSumOptimized(self, nums, target):
        count = {0: 1}
        
        cur_sum = 0
        smaller_count = 0
        ans = 0

        for num in nums:
            prev_sum = cur_sum
            if num == target:
                cur_sum += 1
                smaller_count += count.get(prev_sum, 0)
            else:
                cur_sum -= 1
                smaller_count -= count.get(cur_sum, 0)
            
            ans += smaller_count

            count[cur_sum] = count.get(cur_sum, 0) + 1
        
        return ans



        
```
