# **SOLUTION**

```python
class Solution(object):
    def solveSudoku(self, board):
        rows = [0] * 9
        cols = [0] * 9
        boxes = [0] * 9

        def box_id(r, c):
            return (r // 3) * 3 + (c // 3)

        FULL = 0x3FE  # bits 1..9

        empties = []
        for r in range(9):
            for c in range(9):
                ch = board[r][c]
                if ch == '.':
                    empties.append((r, c))
                else:
                    d = ord(ch) - 48  # '0' -> 48
                    bit = 1 << d
                    b = box_id(r, c)
                    rows[r] |= bit
                    cols[c] |= bit
                    boxes[b] |= bit

        # thay cho cand.bit_count()
        def popcount(x):
            cnt = 0
            while x:
                x &= x - 1
                cnt += 1
            return cnt

        def pick_cell():
            best_i = -1
            best_mask = 0
            best_cnt = 10
            for i, (r, c) in enumerate(empties):
                if r == -1:
                    continue
                b = box_id(r, c)
                cand = (~(rows[r] | cols[c] | boxes[b])) & FULL
                cnt = popcount(cand)       # <— dùng popcount thay vì .bit_count()
                if cnt < best_cnt:
                    best_cnt = cnt
                    best_mask = cand
                    best_i = i
                    if cnt == 1:
                        break
            return best_i, best_mask

        def dfs(left):
            if left == 0:
                return True

            idx, cand = pick_cell()
            if idx == -1 or cand == 0:
                return False

            r, c = empties[idx]
            b = box_id(r, c)
            empties[idx] = (-1, -1)

            m = cand
            while m:
                bit = m & -m
                d = bit.bit_length() - 1   # xác định chữ số 1..9
                ch = chr(d + 48)

                board[r][c] = ch
                rows[r] |= bit
                cols[c] |= bit
                boxes[b] |= bit

                if dfs(left - 1):
                    return True

                board[r][c] = '.'
                rows[r] ^= bit
                cols[c] ^= bit
                boxes[b] ^= bit

                m ^= bit

            empties[idx] = (r, c)
            return False

        dfs(len(empties))
```

```python
solution = Solution()

def run_test_case(case_number, input_board, expected):
    result = solution.solveSudoku(input_board)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - board: {input_board}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [["5","3",".",".","7",".",".",".","."],
         ["6",".",".","1","9","5",".",".","."],
         [".","9","8",".",".",".",".","6","."],
         ["8",".",".",".","6",".",".",".","3"],
         ["4",".",".","8",".","3",".",".","1"],
         ["7",".",".",".","2",".",".",".","6"],
         [".","6",".",".",".",".","2","8","."],
         [".",".",".","4","1","9",".",".","5"],
         [".",".",".",".","8",".",".","7","9"]], 
         [["5","3","4","6","7","8","9","1","2"],
          ["6","7","2","1","9","5","3","4","8"],
          ["1","9","8","3","4","2","5","6","7"],
          ["8","5","9","7","6","1","4","2","3"],
          ["4","2","6","8","5","3","7","9","1"],
          ["7","1","3","9","2","4","8","5","6"],
          ["9","6","1","5","3","7","2","8","4"],
          ["2","8","7","4","1","9","6","3","5"],
          ["3","4","5","2","8","6","1","7","9"]])
]

for case_number, input_board, expected in test_cases:
    run_test_case(case_number, input_board, expected)
```
