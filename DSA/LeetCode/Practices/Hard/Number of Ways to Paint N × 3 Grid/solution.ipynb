# **SOLUTION**

```python
class Solution(object):
    def numOfWays(self, n):
        """
        :type n: int
        :rtype: int
        """
        MOD = 10**9 + 7

        return self.solverMath(n, MOD)
    
    def solverMath(self, n, MOD):
        if n == 1:
            return 12
        if n == 2:
            return 54
        t1, t2 = 12, 54
        for _ in range(3, n + 1):
            t3 = (5 * t2 - 2 * t1 ) % MOD
            t1, t2 = t2, t3
        return t3
```

```python
solution = Solution()

# def extraCondition(result, expected):
#     pass


def run_test_case(case_number, input_n, expected):
    result = solution.numOfWays(input_n)
    status = "PASSED" if result == expected else "FAILED"
    # status = "PASSED" if extraCondition(result, expected) else "FAILED"
    print(f"Case {case_number} - (n): {(input_n)}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, 1, 12),
    (2, 5000, 30228214),
]

for case_number, input_n, expected in test_cases:
    run_test_case(case_number, input_n, expected)
```
