# **SOLUTION**

```python
class Solution(object):
    def numberOfPaths(self, grid, k):
        """
        :type grid: List[List[int]]
        :type k: int
        :rtype: int
        """
        MOD = 10 ** 9 + 7

        return self.solverDP(grid, k, MOD)
    def solverDP(self, grid, k, MOD):
        m, n = len(grid), len(grid[0])
        # dp[i][j][r] = number of paths reaching cell (i, j) with a total sum % k = r
        dp = [[[0] * k for _ in range(n)] for _ in range(m)]

        dp[0][0][grid[0][0] % k] = 1

        for i in range(m):
            for j in range(n):
                if i == 0 and j == 0:
                    continue
                val_mod = grid[i][j] % k
                
                for r in range(k):
                    if i > 0 and dp[i - 1][j][r] > 0:
                        new_r = (r + val_mod) % k
                        dp[i][j][new_r] = (dp[i][j][new_r] + dp[i - 1][j][r]) % MOD

                    if j > 0 and dp[i][j - 1][r] > 0:
                        new_r = (r + val_mod) % k
                        dp[i][j][new_r] = (dp[i][j][new_r] + dp[i][j - 1][r]) % MOD

        return dp[m - 1][n - 1][0]

        
```
