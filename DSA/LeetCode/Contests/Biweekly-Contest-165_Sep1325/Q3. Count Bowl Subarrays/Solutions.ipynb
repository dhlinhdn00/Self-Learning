# **SOLUTION**

```python
class Solution(object):
    def bowlSubarrays(self, nums):
        return self.solver(nums)

    def solver(self, nums):
        n = len(nums)
        if n < 3:
            return 0

        # Nearest Greater to Left (L)
        L = [-1] * n
        st = []
        for i in range(n):
            while st and nums[st[-1]] <= nums[i]:
                st.pop()
            L[i] = st[-1] if st else -1
            st.append(i)

        # Nearest Greater to Right (R)
        R = [-1] * n
        st = []
        for i in range(n - 1, -1, -1):
            while st and nums[st[-1]] <= nums[i]:
                st.pop()
            R[i] = st[-1] if st else -1
            st.append(i)

        ans = 0
        for j in range(1, n - 1):
            if L[j] != -1 and R[j] != -1:
                ans += 1

        return ans
```

# **DRAFT**

```python
nums = [2,5,3,1,4]
n = len(nums)

L = [-1] * n
st = []
for i in range(n):
    while st and nums[st[-1]] <= nums[i]:
        st.pop()
    L[i] = st[-1] if st else -1
    st.append(i)
    # print(L)

R = [-1] * n
st.clear()
for i in range(n - 1, -1, -1):
    while st and nums[st[-1]] <= nums[i]:
        st.pop()

    R[i] = st[-1] if st else -1
    st.append(i)

    ans = 0
    for i in range(1, n-1):
        if L[i] != -1 and R[i] != -1:
            ans += 1
print(ans)
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_nums, expected):
    result = solution.bowlSubarrays(input_nums)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - nums: {input_nums}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [2,5,3,1,4], 2),
    (2, [5,1,2,3,4], 3),
    (3, [1000000000,999999999,999999998], 0)
]

for case_number, input_s, expected in test_cases:
    run_test_case(case_number, input_s, expected)
```
