# **SOLUTION**

```python
class Solution(object):
    def minOperations(self, s):
        """
        :type s: str
        :rtype: int
        """
        return self.solverNaive(s)

    def solverNaive(self, s):
        arr = set([ord(c) for c in s]) - set({97})
        n = len(arr)
        if n == 0:
            return 0
        arr = sorted(arr)

        ans = 1

        for i in range(1, n):
            ans += (arr[i] - arr[i - 1])

        if 122 not in arr:
            ans += 122 - arr[n - 1]

        return ans
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_s, expected):
    result = solution.minOperations(input_s)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - s: {input_s}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, "yz", 2),
    (2, "a", 0),
]

for case_number, input_s, expected in test_cases:
    run_test_case(case_number, input_s, expected)
```
