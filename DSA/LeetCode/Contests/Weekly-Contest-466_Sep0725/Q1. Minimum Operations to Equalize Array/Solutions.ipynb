# **SOLUTION**

```python
class Solution(object):
    def minOperations(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        return self.solverNaive(nums)

    def solverNaive(self, nums):
        if all(x == nums[0] for x in nums):
            return 0
        return 1
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_word, expected):
    result = solution.minOperations(input_word)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - word: {input_word}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [1, 2], 1),
    (2, [5, 5, 5], 0),
]

for case_number, input_word, expected in test_cases:
    run_test_case(case_number, input_word, expected)
```
