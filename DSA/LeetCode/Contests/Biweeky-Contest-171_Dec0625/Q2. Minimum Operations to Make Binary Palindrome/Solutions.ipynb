# **DRAFT**

```python
import math
def isPrime(num):
    if num < 2:
        return False
    if num % 2 == 0:
        return num == 2
    limit = int(math.sqrt(num))
    for i in range(3, limit + 1, 2):
        if num % i == 0:
            return False
        return True

num = 5237
n = len(str(num))
for i in range(n):
    prefix = int(num / 10**i)
    suffix = int(str(int(int(str(num)[::-1]) // 10**i))[::-1])
    if isPrime(prefix) == False or isPrime(suffix) == False:
        print(False)
```
