# **SOLUTION**

```python
class Solution(object):
    def maximumLength(self, nums):
        """
        Given an array nums, return the length of the longest subsequence such that
        for every adjacent pair (sub[i], sub[i+1]) the parity of their sum is constant
        across the whole subsequence.

        We'll build subsequences greedily in the spirit of the user's original code:
        - pick a starting element
        - keep a running 'prev' (last accepted value)
        - scan forward; accept nums[i] iff (prev + nums[i]) % 2 == parity_target
        This is optimal for a fixed (parity_target, start_parity) because skipping
        elements never reduces future options.

        We try four scenarios:
            parity_target = 0 (every pair sums to even) starting from first even
            parity_target = 0 starting from first odd
            parity_target = 1 (every pair sums to odd)  starting from first even
            parity_target = 1 starting from first odd

        The answer is the maximum length over these four greedy passes.
        """
        def build(parity_target, start_parity):
            """
            parity_target: 0 => want (prev + cur) even every time; 1 => want odd every time.
            start_parity:  0 => first taken element must be even; 1 => must be odd.

            Returns length of the longest subsequence obtainable under these constraints,
            respecting original order (greedy is optimal here).
            """
            # Find first element whose parity matches start_parity.
            i0 = None
            for i, x in enumerate(nums):
                if (x & 1) == start_parity:
                    i0 = i
                    break
            if i0 is None:
                return 0  # no element of required start parity exists

            count = 1
            prev = nums[i0]

            # Scan forward; only take elements that keep the pair-sum parity constant.
            for x in nums[i0 + 1:]:
                if ((prev + x) & 1) == parity_target:
                    count += 1
                    prev = x  # advance subsequence tail

            return count

            # Note: We do NOT update parity_target while scanning.
            # Keeping it fixed enforces "all pair sums have the same parity" correctly.
            # Alternation emerges naturally when parity_target == 1, because each accepted
            # element must differ in parity from 'prev'.

        # Four scenarios; take the best.
        same_from_even = build(0, 0)  # all-even-sum, start at an even
        same_from_odd  = build(0, 1)  # all-even-sum, start at an odd
        alt_from_even  = build(1, 0)  # all-odd-sum  , start at an even
        alt_from_odd   = build(1, 1)  # all-odd-sum  , start at an odd

        return max(same_from_even, same_from_odd, alt_from_even, alt_from_odd)
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_nums, expected):
    result = solution.maximumLength(input_nums)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - nums: {input_nums}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [1,2,3,4], 4),
    (2, [1,2,1,1,2,1,2], 6),
    (3, [1,3], 2)
]

for case_number, input_nums, expected in test_cases:
    run_test_case(case_number, input_nums, expected)
```
