# **SOLUTION**

```python
class Solution(object):
    def countSquares(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: int
        """
        return self.solverDP(matrix)
    
    def solverNaive(self, matrix):
        """
        :method: Brute-force
        :complexity: TLE
        """
        h = len(matrix)
        w = len(matrix[0])
        count = 0
        
        for size in range(1, min(h, w) + 1):
            for i in range(h - size + 1):
                for j in range(w - size + 1):
                    sub = []
                    for row in matrix[i : i + size]:
                        sub.append(row[j : j + size])
                    if all(all(x == 1 for x in row) for row in sub):
                        count += 1
        return count
    
    def solverDP(self, matrix):
        h = len(matrix)
        w = len(matrix[0])
        count = 0
        dp = [[0] * w for _ in range(h)]
        
        for i in range(h):
            for j in range(w):
                if matrix[i][j] == 1:
                    if i == 0 or j == 0: # margin: cannot create more than 2 sub-square matrices
                        dp[i][j] = 1
                    else:
                        dp[i][j] = 1 + min(dp[i - 1][j], # top
                                        dp[i][j - 1], # left
                                        dp[i - 1][j - 1] #top -left
                                        )
                    count += dp[i][j]
        return count
```

# **DRAFT**

```python
# Learn yield. Function has yield as return => called generator function
def demo_return(n):
    result = []
    for i in range(n):
        result.append(i)
    return result   # all for one
def demo_yield(n):
    for i in range(n):
        yield i     # gradually

print("return:", demo_return(5))
print("yield:")
for x in demo_yield(5):
    print(x)
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_matrix, expected):
    result = solution.countSquares(input_matrix)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - matrix: {input_matrix}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [[1,0,1], [1,1,0], [1,1,0]], 7),
    (2, [[0,1,1,1], [1,1,1,1], [0,1,1,1]], 15),
]

for case_number, input_matrix, expected in test_cases:
    run_test_case(case_number, input_matrix, expected)
```
