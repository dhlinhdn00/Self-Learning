# **DRAFT**

```python
fruits = [4, 2, 5]
baskets = [3, 5, 4]
completion_count = 0

for fruit in fruits:
    for basket in baskets:
        if fruit <= basket:
            completion_count += 1
            baskets.remove(basket)
            break

ans = len(fruits) - completion_count

print(ans)
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_fruits, input_baskets, expected):
    result = solution.numOfUnplacedFruits(input_fruits, input_baskets)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - fruits: {input_fruits}, baskets: {input_baskets}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [4, 2, 5], [3, 5, 4], 1),
    (2, [3, 6, 1], [6, 4, 7], 0),
]

for case_number, input_fruits, input_baskets, expected in test_cases:
    run_test_case(case_number, input_fruits, input_baskets, expected)
```
