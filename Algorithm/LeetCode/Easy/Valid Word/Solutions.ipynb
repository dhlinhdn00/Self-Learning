# **SOLUTION**

```python
class Solution(object):
    def isValid(self, word):
        """
        :type word: str
        :rtype: bool
        """
        return self.solverNaive(word)
    
    def solverNaive(self, word):
        if len(word) < 3:
            return False
        vowels = set("aeiouAEIOU")
        has_vowel = False
        has_consonant = False
        
        for ch in word:
            ch_code = ord(ch)
            if not (
                (48 <= ch_code <= 57) or
                (65 <= ch_code <= 90) or
                (97 <= ch_code <= 122)
            ):
                return False
            if (65 <= ch_code <= 90) or (97 <= ch_code <= 122):
                if ch in vowels:
                    has_vowel = True
                else:
                    has_consonant = True

        return has_vowel and has_consonant
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_word, expected):
    result = solution.isValid(input_word)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - word: {input_word}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, "234Adas", True),
    (2, "b3", False),
    (3, "a3$e", False)
]

for case_number, input_word, expected in test_cases:
    run_test_case(case_number, input_word, expected)
```
