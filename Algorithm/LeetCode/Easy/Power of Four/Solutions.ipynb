# **SOLUTION**

```python
class Solution(object):
    def isPowerOfFour(self, n):
        """
        :type n: int
        :rtype: bool
        """
        return self.solverNaive(n)

    def solverNaive(self, n):
        if n <= 0:
            return False
        while n % 4 == 0:
            n //= 4
        return n == 1
        
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_n, expected):
    result = solution.isPowerOfFour(input_n)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - n: {input_n}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, 16, True),
    (2, 5, False),
    (3, 1, True)
]

for case_number, input_fruits, input_baskets, expected in test_cases:
    run_test_case(case_number, input_fruits, input_baskets, expected)
```
