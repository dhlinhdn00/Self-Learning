# **SOLUTION**

```python
class Solution(object):
    def maximum69Number (self, num):
        """
        :type num: int
        :rtype: int
        :constraint: not using String function!
        """
        return self.solver69(num)

    def solver69(self, num):
        x = num
        pos = -1
        i = 0
        while x > 0:
            if x % 10 == 6:
                pos = i
            x //= 10
            i += 1
        if pos == -1:
            return num
        return num + 3 * (10**pos)
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_num, expected):
    result = solution.maximum69Number(input_num)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - num: {input_num}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, 9669, 9969),
    (2, 9996, 9999),
    (3, 9999, 9999)
]

for case_number, input_num, expected in test_cases:
    run_test_case(case_number, input_num, expected)
```
