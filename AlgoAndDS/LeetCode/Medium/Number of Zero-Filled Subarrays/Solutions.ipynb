# **SOLUTION**

```python
class Solution(object):
    def zeroFilledSubarray(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        return self.solverDP(nums)
    
    def solverDP(self, nums):
        n = len(nums)
        dp = [0] * n
        count = 0

        for i in range(n):
            if nums[i] == 0:
                if i == 0:
                    dp[i] = 1
                else:
                    dp[i] = 1 + dp[i - 1]

                count += dp[i]

        return count
        
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_nums, expected):
    result = solution.zeroFilledSubarray(input_nums)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - nums: {input_nums}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [1,3,0,0,2,0,0,4], 6),
    (2, [0,0,0,2,0,0], 9),
    (3, [2,10,2019], 0)
]

for case_number, input_nums, expected in test_cases:
    run_test_case(case_number, input_nums, expected)
```
