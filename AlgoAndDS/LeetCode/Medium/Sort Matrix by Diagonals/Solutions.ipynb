# **SOLUTION**

```python
class Solution(object):
    def sortMatrix(self, grid):
        """
        :type grid: List[List[int]] 
        :rtype: List[List[int]]
        """
        return self.solverNaive(grid)
    
    def solverNaive(self, grid):
        h, w = len(grid), len(grid[0])
        for col in range(w - 1, 0 , -1): # top right
            i, j = 0, col
            diag = []
            while i < h and j < w:
                diag.append(grid[i][j])
                i += 1
                j += 1
            diag.sort()
            
            i, j = 0, col
            k = 0
            while i < h and j < w:
                grid[i][j] = diag[k]
                i += 1
                j += 1
                k += 1

        for row in range(h): # bot, left
            i, j = row, 0
            diag = []
            while i < h and j < w:
                diag.append(grid[i][j])
                i += 1
                j += 1
            diag.sort(reverse=True)

            i, j = row, 0
            k = 0
            while i < h and j < w:
                grid[i][j] = diag[k]
                i += 1
                j += 1
                k += 1
        return grid
```

# **DRAFT**

```python
grid = [[1, 7, 3],
        [9, 8, 2],
        [4, 5, 6]]

h, w = len(grid), len(grid[0])

for col in range(w - 1, 0, -1):
    i, j = 0, col
    diag = []

    while i < h and j < w:
        diag.append(grid[i][j])
        i += 1
        j += 1

    diag.sort()

    i, j = 0, col
    k = 0
    while i < h and j < w:
        grid[i][j] = diag[k]
        i += 1
        j += 1
        k += 1

for row in range(h):
    i, j = row, 0
    diag = []

    while i < h and j < w:
        diag.append(grid[i][j])
        i += 1
        j += 1

    diag.sort(reverse=True)

    i, j = row, 0
    k = 0
    while i < h and j < w:
        grid[i][j] = diag[k]
        i += 1
        j += 1
        k += 1

for r in grid:
    print(r)
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_grid, expected):
    result = solution.sortMatrix(input_grid)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - grid: {input_grid}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [[1,7,3],[9,8,2],[4,5,6]], [[8,2,3],[9,6,7],[4,5,1]]),
    (2, [[0,1],[1,2]], [[2,1],[1,0]]),
    (3, [[1]], [[1]])
]

for case_number, input_grid, expected in test_cases:
    run_test_case(case_number, input_grid, expected)
```
