# **SOLUTION**

```python
class Solution(object):
    def maximumLength(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        return self.solverGreedy(nums)
    
    def solverGreedy(self, nums):
        def build(start, surplus, nums):
            """
            :start: starting num is even or odd ? => 0 stands for even | 1 stands for odd
            :surplus: even or odd ? => 0 stands for even | 1 stands for odd
            """
            start_pos = None
            n = len(nums)
            for i in range(n):
                if nums[i] % 2 == start:
                    start_pos = i
                    break
            if start_pos == None:
                return 0
            
            count = 1
            prev = nums[start_pos]
            for i in range(start_pos + 1, n):
                if (prev + nums[i]) % 2 == surplus:
                    count += 1
                    prev = nums[i]
            return count
        
        even_start_even_surplus = build(0, 0, nums)
        even_start_odd_surplus = build(0, 1, nums)
        odd_start_odd_surplus = build(1, 1, nums)
        odd_start_even_surplus = build(1, 0, nums)
        return max(even_start_even_surplus, even_start_odd_surplus, odd_start_odd_surplus, odd_start_even_surplus)

        
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_nums, expected):
    result = solution.maximumLength(input_nums)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - nums: {input_nums}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [1,2,3,4], 4),
    (2, [1,2,1,1,2,1,2], 6),
    (3, [1,3], 2)
]

for case_number, input_nums, expected in test_cases:
    run_test_case(case_number, input_nums, expected)
```
