# **SOLUTION**

```python
class Solution(object):
    def isValidSudoku(self, board):
        """
        :type board: List[List[str]]
        :rtype: bool
        """
        return self.solverNaive(board)

    def solverNaive(self, board):
        cols = [set() for _ in range(9)]
        rows = [set() for _ in range(9)]
        blocks = [set() for _ in range(9)]

        for c in range(9):
            for r in range(9):
                val = board[r][c]
                if val == ".":
                    continue

                b = (r // 3) * 3 + (c // 3)

                if (val in rows[r]) or (val in cols[c]) or (val in blocks[b]):
                    return False
                
                cols[c].add(val)
                rows[r].add(val)
                blocks[b].add(val)

        return True
```
