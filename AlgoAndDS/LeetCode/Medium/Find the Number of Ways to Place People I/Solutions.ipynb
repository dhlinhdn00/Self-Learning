# **SOLUTION**

```python
class Solution(object):
    def numberOfPairs(self, points):
        """
        :type points: List[List[int]]
        :rtype: int
        """
        return self.solverNaive(points)

    def solverNaive(self, points):
        n = len(points)
        ans = 0 

        for i in range(n):
            xA, yA = points[i]
            for j in range(n):
                if i == j:
                    continue
                xB, yB = points[j]

                # upper left condition
                if xA > xB or yB > yA:
                    continue
                
                for k in range(n):
                    if k == i or k == j:
                        continue
                    xC, yC = points[k]
                    
                    # no other points in the rectangle
                    if xA <= xC <= xB and yB <= yC <= yA:
                        break
    
                else:
                    ans += 1

        return ans
                
```

# **DRAFT**

```python
points = [[0,0],[0,6],[0,3],[1,0]]

n = len(points)
ans = 0 

for i in range(n):
    xA, yA = points[i]
    for j in range(n):
        if i == j:
            continue
        xB, yB = points[j]
        # upper left condition
        if xA > xB or yB > yA:
            continue

        for k in range(n):
            if k == i or k == j:
                continue
            xC, yC = points[k]

            if xA <= xC <= xB and yB <= yC <= yA:
                continue

            ans += 1
print(ans)
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_points, expected):
    result = solution.numberOfPairs(input_points)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - points: {input_points}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [[1,1],[2,2],[3,3]], 0),
    (2, [[6,2],[4,4],[2,6]], 2),
    (3, [[3,1],[1,3],[1,1]], 2)
]

for case_number, input_points, expected in test_cases:
    run_test_case(case_number, input_points, expected)
```
