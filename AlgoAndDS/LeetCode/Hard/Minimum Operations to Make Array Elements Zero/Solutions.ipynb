# **SOLUTION**

```python
class Solution(object):
    def minOperations(self, queries):
        """
        :type queries: List[List[int]]
        :rtype: int
        """

        # t(x) = số lần cần để x về 0
        def t_of(x):
            if x <= 0:
                return 0
            k, p = 0, 1  # p = 4^k
            while p <= x:
                p *= 4
                k += 1
            return k

        # prefix sum S(N) = sum_{i=1..N} t(i)
        def prefix_sum_t(N):
            if N <= 0:
                return 0
            S = 0
            p = 1   # 4^0
            k = 1
            while p <= N:
                upper = min(N, 4*p - 1)
                cnt = upper - p + 1
                if cnt > 0:
                    S += k * cnt
                p *= 4
                k += 1
            return S

        def ops_for_range(l, r):
            T = prefix_sum_t(r) - prefix_sum_t(l - 1)
            M = t_of(r)
            return max(M, (T + 1) // 2)

        total = 0
        for l, r in queries:
            total += ops_for_range(l, r)
        return total
```

```python
solution = Solution()

def run_test_case(case_number, input_queries, expected):
    result = solution.minOperations(input_queries)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - queries: {input_queries}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [[1,2],[2,4]], 3),
    (2, [[2,6]], 4),
]

for case_number, input_queries, expected in test_cases:
    run_test_case(case_number, input_queries, expected)
```
