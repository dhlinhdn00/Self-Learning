# **SOLUTION**

```python
class Solution(object):
    def smallestAbsent(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        return self.solverNaive(nums)
    def solverNaive(self, nums):
        positive_set = set(x for x in nums if x > 0)
        s = sum(nums)
        n = len(nums)
        avg = s // n + 1
        if avg <= 0:
            avg = 1
        while avg in positive_set:
            avg += 1
        return int(avg)
```

# **DRAFT**

```python
nums = [3, 5]
s = sum(nums)
n = len(nums)
avg = round(s / n) if s % n != 0 else round(s / n) + 1
while avg in nums:
    avg += 1
avg
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_nums, expected):
    result = solution.smallestAbsent(input_nums)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - nums: {input_nums}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [3, 5], 6),
    (2, [-1,1,2], 3),
    (3, [4,-1], 2)
]

for case_number, input_nums, expected in test_cases:
    run_test_case(case_number, input_nums, expected)
```
