# **SOLUTION**

```python
class Solution(object):
    def findClosest(self, x, y, z):
        """
        :type x: int
        :type y: int
        :type z: int
        :rtype: int
        """
        return self.solverNaive(x, y, z)

    def solverNaive(self, x, y, z):
        dx = abs(x - z)
        dy = abs(y -z)
        return 2 if dx > dy else 1 if dx < dy else 0
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_x, input_y, input_z, expected):
    result = solution.findClosest(input_x, input_y, input_z)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - x, y, z: {input_x, input_y, input_z}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, 2, 5, 6, 2),
    (2, 1, 5, 3, 0),
    (3, 2, 7, 4, 1)
]

for case_number, input_x, input_y, input_z, expected in test_cases:
    run_test_case(case_number, input_x, input_y, input_z, expected)
```
