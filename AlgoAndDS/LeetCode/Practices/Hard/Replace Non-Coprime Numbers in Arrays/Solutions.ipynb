# **SOLUTION**

```python
class Solution(object):
    def replaceNonCoprimes(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        return self.solver(nums)
    
    def gcd(self, a, b):
        while b:
            a, b = b, a % b
        return a

    def solver(self, nums):
        st = []
        for x in nums:
            cur = x
            while st:
                g = self.gcd(st[-1], cur)
                if g == 1:
                    break
                cur = (st[-1] // g) * cur
                st.pop()
            st.append(cur)
        return st
```

```python
solution = Solution()

def run_test_case(case_number, input_nums, expected):
    result = solution.replaceNonCoprimes(input_nums)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - nums: {input_nums}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [6,4,3,2,7,6,2], [12, 7, 6]),
    (2, [2,2,1,1,3,3,3], [2, 1, 1, 3])
]

for case_number, input_nums, expected in test_cases:
    run_test_case(case_number, input_nums, expected)
```
