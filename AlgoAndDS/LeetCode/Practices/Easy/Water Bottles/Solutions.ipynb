# **SOLUTION**

```python
class Solution(object):
    def numWaterBottles(self, numBottles, numExchange):
        """
        :type numBottles: int
        :type numExchange: int
        :rtype: int
        """
        return self.solverNaive(numBottles, numExchange)

    def solverNaive(self, numBottles, numExchange):
        ans = numBottles
        r = float("inf")
        while numBottles >= numExchange:
            numBottles, r = divmod(numBottles, numExchange)
            ans += numBottles
            numBottles += r
        return ans
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_numBottles, input_numExchange, expected):
    result = solution.numWaterBottles(input_numBottles, input_numExchange)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - (numBottles, numExchange): ({input_numBottles, input_numExchange}), Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, 9, 3, 13),
    (2, 15, 4, 19),
]

for case_number, input_numBottles, input_numExchange, expected in test_cases:
    run_test_case(case_number, input_numBottles, input_numExchange, expected)
```
