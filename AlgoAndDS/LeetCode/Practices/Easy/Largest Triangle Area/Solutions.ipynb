# **SOLUTION**

```python
class Solution(object):
    def largestTriangleArea(self, points):
        """
        :type points: List[List[int]]
        :rtype: float
        """
        return self.solverNaive(points)

    def solverNaive(self, points):
        def ShoelaceFormulaArea(p1, p2, p3):
            return abs(p1[0]*(p2[1] - p3[1]) + p2[0]*(p3[1] - p1[1]) + p3[0]*(p1[1] - p2[1])) / 2.0
        max_area = 0
        for p1 in points:
            for p2 in points:
                for p3 in points:
                    if p1 != p2 and p2 != p3 and p1 != p3:
                        area = ShoelaceFormulaArea(p1, p2, p3)
                        max_area = max(area, max_area)
        
        return round(max_area, 5)

        
```
