# **SOLUTION**

```python
import math

class Solution(object):
    def sumZero(self, n):
        """
        :type n: int
        :rtype: List[int]
        """
        return self.solverNaive(n)

    def solverNaive(self, n):
        L = []
        start_n = int(math.floor(n // 2))
        for i in range(start_n):
            L.append(i + 1)
            L.append((i + 1) * (-1))
        if n % 2 != 0:
            L.append(0)
        return L
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_n):
    result = solution.sumZero(input_n)
    status = "PASSED" if sum(result) == 0 else "FAILED"
    print(f"Case {case_number} - n: {input_n}, Output: {result}, Sum of Output: {sum(result)}, Status: {status}")

test_cases = [
    (1, 5),
    (2, 3),
    (3, 1)
]

for case_number, input_n in test_cases:
    run_test_case(case_number, input_n)
```
