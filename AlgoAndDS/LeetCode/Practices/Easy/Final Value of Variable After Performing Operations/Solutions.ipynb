# **SOLUTION**

```python
class Solution(object):
    def finalValueAfterOperations(self, operations):
        """
        :type operations: List[str]
        :rtype: int
        """
        return self.solverNaive(operations)

    def solverNaive(self, operations):
        for i in range(len(operations)):
            state = operations[i]
            if state == "X++" or state == "++X":
                operations[i] = 1
            elif state == "X--" or state == "--X":
                operations[i] = -1

        return sum(operations)        
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_operations, expected):
    result = solution.finalValueAfterOperations(input_operations)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - operations: {input_operations}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, ["--X","X++","X++"], 1),
    (2, ["++X","++X","X++"], 3),
    (3, ["X++","++X","--X","X--"], 0)
]

for case_number, input_operations, expected in test_cases:
    run_test_case(case_number, input_operations, expected)
```
