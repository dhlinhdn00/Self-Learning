# **SOLUTION**

```python
class Solution(object):
    def getNoZeroIntegers(self, n):
        """
        :type n: int
        :rtype: List[int]
        """
        return self.solverNaive(n)

    def solverNaive(self, n):
        for i in range(n):
            m = n - i
            if "0" not in str(i) and "0" not in str(m):
                return [m, i]
        
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_n):
    result = solution.getNoZeroIntegers(input_n)
    status = "PASSED"
    for num in result:
        if "0" in str(num):
            status = "FAILED"
    print(f"Case {case_number} - word: {input_n}, Output: {result}, Status: {status}")

test_cases = [
    (1, 2),
    (2, 11)
]

for case_number, input_n in test_cases:
    run_test_case(case_number, input_n)
```
