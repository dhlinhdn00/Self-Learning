# **SOLUTION**

```python
class Solution(object):
    def canBeTypedWords(self, text, brokenLetters):
        """
        :type text: str
        :type brokenLetters: str
        :rtype: int
        """
        return self.solverNaive(text, brokenLetters)

    def solverNaive(self, text, brokenLetters):
        wordList = text.split()
        n = len(wordList)
        ans = n
        for word in wordList:
            for c in brokenLetters:
                if c in word:
                    ans -= 1
                    break
        return ans
    
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_text, input_brokenLetters, expected):
    result = solution.canBeTypedWords(input_text, input_brokenLetters)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - (text, brokenLetters): {input_text, input_brokenLetters}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, "hello world", "ad", 1),
    (2, "leet code", "lt", 1),
    (3, "leet code", "e", 0)
]

for case_number, input_text, input_brokenLetters, expected in test_cases:
    run_test_case(case_number, input_text, input_brokenLetters, expected)
```
