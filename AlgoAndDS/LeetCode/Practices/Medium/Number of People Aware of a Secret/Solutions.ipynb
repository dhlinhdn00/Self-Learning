# **SOLUTION**

```python
class Solution(object):
    def peopleAwareOfSecret(self, n, delay, forget):
        """
        :type n: int
        :type delay: int
        :type forget: int
        :rtype: int
        """
        MOD = 10**9 + 7
        return self.solverWin(n, delay, forget, MOD)

    def solverWin(self, n, d, f, MOD):
        new, new[1] = [0] * (n + 1), 1
        known, known[1] = [0] * (n + 1), 1
        S = 0 # the number of people able to share secret on day i

        for i in range(2, n + 1):
            S = (S + (new[i - d] if i - d > 0 else 0) - (new[i - f] if i - f > 0 else 0)) % MOD
            new[i] = S
            forgot = new[i - f] if i - f > 0 else 0
            known[i] = (known[i - 1] + new[i] - forgot) % MOD

        return known[n] % MOD

```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_n, input_delay, input_forget, expected):
    result = solution.peopleAwareOfSecret(input_n, input_delay, input_forget)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - (n, delay, forget): {input_n, input_delay, input_forget}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, 6, 2, 4, 5),
    (2, 4, 1, 3, 6),
    (3, 6, 1, 2, 2),
    (4, 5, 1, 3, 10)
]

for case_number, input_n, input_delay, input_forget, expected in test_cases:
    run_test_case(case_number, input_n, input_delay, input_forget, expected)
```
