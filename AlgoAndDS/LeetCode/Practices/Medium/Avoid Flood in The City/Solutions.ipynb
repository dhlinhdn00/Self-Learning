# **SOLUTION**

```python
import bisect

class Solution(object):
    def avoidFlood(self, rains):
        """
        :type rains: List[int]
        :rtype: List[int]
        """
        return self.solverNaive(rains)
    
    def solverNaive(self, rains):
        n = len(rains)
        full = {}          # {lake: last_rain_day}
        dry_days = []      # list of days we can dry
        ans = [-1] * n

        for i, lake in enumerate(rains):
            if lake == 0:
                dry_days.append(i)
                ans[i] = 1   
            else:
                if lake in full:
                    prev_rain = full[lake]
                    idx = bisect.bisect_right(dry_days, prev_rain)
                    if idx == len(dry_days):
                        return []  
                    dry_day = dry_days[idx]
                    ans[dry_day] = lake
                    dry_days.pop(idx)
                full[lake] = i
                ans[i] = -1

        return ans
            
```

# **DRAFT**

```python
from collections import deque

rains = [1,2,0,0,2,1]
n = len(rains)
ans = []
full_lakes = deque()
for i in range(n):
    status = rains[0]
    if status != 0:
        ans.append(-1)
        if status in full_lakes:
            return []
        else:
            full_lakes.appendleft(status)
    else:
        dried_lake = full_lakes.pop()
        ans.append(dried_lake)
return ans
    

    


ans
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_rains, expected):
    result = solution.avoidFlood(input_rains)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - rains: {input_rains}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [1,2,3,4], [-1,-1,-1,-1]),
    (2, [1,2,0,0,2,1], [-1,-1,2,1,-1,-1]),
    (3, [1,2,0,1,2], [])
]

for case_number, input_rains, expected in test_cases:
    run_test_case(case_number, input_rains, expected)
```
