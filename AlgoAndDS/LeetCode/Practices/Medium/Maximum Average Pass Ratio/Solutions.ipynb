# **SOLUTION**

```python
class Solution(object):
    def maxAverageRatio(self, classes, extraStudents):
        """
        :type classes: List[List[int]]
        :type extraStudents: int
        :rtype: float
        """
        return self.solverGreedyUsingPriorityQueue(classes, extraStudents)

    def solverGreedyUsingList(self, classes, extraStudents):
        """
        cost: O(n * extraStudents) => TLE"
        """
        n = len(classes)
        p = [c[0] for c in classes] # passed-list
        t = [c[1] for c in classes] # total-list

        for _ in range(extraStudents):
            best_g = 0 # best marginal gain
            best_i = 0 # index of best marginal gain

            for i in range(n):
                # gain = (p + 1) / (t + 1) - p / t = (t - p) / t * (t + 1)
                g = (t[i] - p[i]) / float(t[i] * (t[i] + 1))

                if g > best_g:
                    best_g = g
                    best_i = i

            p[best_i] += 1
            t[best_i] += 1

        return round(sum(pi / float(ti) for pi, ti in zip(p, t)) / float(n), 5)
    
    def solverGreedyUsingPriorityQueue(self, classes, extraStudents):
        """
        cost: O((n + extraStudents) * log n)
        """
        import heapq

        n = len(classes)
        def gain(p, t):
            return (t - p) / float(t * (t + 1)) 

        ratio_sum = sum(p / float(t) for p, t in classes)

        heap = [(-gain(p, t), p, t) for p, t in classes] # switch -gain to take advantage of min-heap
        heapq.heapify(heap) # switch a list to a heapq

        for _ in range(extraStudents):
            neg_g, p, t = heapq.heappop(heap)  
            ratio_sum += -neg_g            
            p += 1
            t += 1
            heapq.heappush(heap, (-gain(p, t), p, t))

        return round(ratio_sum / n, 5)
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_classes, input_extraStudents, expected):
    result = solution.maxAverageRatio(input_classes, input_extraStudents)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - classes: {input_classes}, extraStudents: {input_extraStudents}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [[1,2],[3,5],[2,2]], 2, 0.78333),
    (2, [[2,4],[3,9],[4,5],[2,10]], 4, 0.53485),
]

for case_number, input_classes, input_extraStudents, expected in test_cases:
    run_test_case(case_number, input_classes, input_extraStudents, expected)
```
