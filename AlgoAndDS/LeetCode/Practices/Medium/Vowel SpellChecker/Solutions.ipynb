# **SOLUTION**

```python
class Solution(object):
    def spellchecker(self, wordlist, queries):
        """
        :type wordlist: List[str]
        :type queries: List[str]
        :rtype: List[str]
        """
        return self.solverNaive(wordlist, queries)
    
    def assign(self, s, i, c):
        return s[:i] + c + s[i + 1:]
    
    def convertWordList(self, l):
        VOWELS = set("AEOUIaeoui")
        lowercased_l = [s.lower() for s in l]
        vowelschanged_l = lowercased_l[:]
        for i in range(len(vowelschanged_l)):
            for j in range(len(vowelschanged_l[i])):
                if vowelschanged_l[i][j] in VOWELS:
                    vowelschanged_l[i] = self.assign(vowelschanged_l[i], j, "-")
        return lowercased_l, vowelschanged_l
    
    def solverNaive(self, wordlist, queries):
        lowercased_wordlist, vowelschanged_wordlist = self.convertWordList(wordlist)
        lowercased_queries, vowelschanged_queries = self.convertWordList(queries)
        
        n = len(queries)
        ans = []
        for i in range(n):
            if queries[i] in wordlist:
                ans.append(queries[i])
            elif lowercased_queries[i] in lowercased_wordlist:
                index = lowercased_wordlist.index(lowercased_queries[i])
                ans.append(wordlist[index])
            elif vowelschanged_queries[i] in vowelschanged_wordlist:
                index = vowelschanged_wordlist.index(vowelschanged_queries[i])
                ans.append(wordlist[index])
            else:
                ans.append("")
        return ans
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_wordlist, input_queries, expected):
    result = solution.spellchecker(input_wordlist, input_queries)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - (wordlist - queries): {input_wordlist, input_queries}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, ["KiTe","kite","hare","Hare"], ["kite","Kite","KiTe","Hare","HARE","Hear","hear","keti","keet","keto"], ["kite","KiTe","KiTe","Hare","hare","","","KiTe","","KiTe"]),
    (2, ["yellow"], ["YellOw"], ["yellow"])
]

for case_number, input_wordlist, input_queries, expected in test_cases:
    run_test_case(case_number, input_wordlist, input_queries, expected)
```
