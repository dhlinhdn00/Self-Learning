# **SOLUTION**

```python
class Solution(object):
    def sortVowels(self, s):
        """
        :type s: str
        :rtype: str
        """
        return self.solverNaive(s)
    
    def assign(self, s, i, c):
        return s[:i] + c + s[i + 1:]

    
    def solverNaive(self, s):
        VOWELS = set("aeiouAEIOU")
        lst_c = []
        lst_i = []
        for i, c in enumerate(s):
            if c in VOWELS:
                lst_c.append(c)
                lst_i.append(i)
        if lst_c == []:
            return s
        
        lst_c = sorted(lst_c)
        for c, i in zip(lst_c, lst_i):
            s = self.assign(s, i, c)
        return s
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_s, expected):
    result = solution.sortVowels(input_s)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - s: {input_s}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, "lEetcOde", "lEOtcede"),
    (2, "lYmpH", "lYmpH")
]

for case_number, input_s, expected in test_cases:
    run_test_case(case_number, input_s, expected)
```
