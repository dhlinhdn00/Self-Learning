# **SOLUTION**

```python
class Solution(object):
    def longestSubarray(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        return self.solverNaive(nums)
    
    def solverNaive(self, nums):
        s = sum(nums)
        n = len(nums)            

        if s == 0:
            return 0
        if n == s:
            return s - 1
        if n <= 3:
            return s

        ans = 0
        dp = n * [0]
        dp[0] = nums[0]
        prev = nums[0]
        for i in range(1, n):
            if nums[i] == 1:
                dp[i] = 1
                if nums[i] == prev:
                    dp[i] += dp[i - 1]
                    dp[i - 1] = -1
            prev = nums[i]

        ck_lst = [x for x in dp if x != -1]

        if len(ck_lst) < 3:
            return max(ck_lst)

        for i in range(0, len(ck_lst) - 2):
            if ck_lst[i:i+3][1] == 0:
                ans = max(sum(ck_lst[i:i+3]), ans)

        return ans
```

# **TESTCASES**

```python
solution = Solution()

def run_test_case(case_number, input_nums, expected):
    result = solution.longestSubarray(input_nums)
    status = "PASSED" if result == expected else "FAILED"
    print(f"Case {case_number} - nums: {input_nums}, Output: {result}, Expected: {expected}, Status: {status}")

test_cases = [
    (1, [1,1,0,1], 3),
    (2, [0,1,1,1,0,1,1,0,1], 5),
    (3, [1,1,1], 2)
]

for case_number, input_nums, expected in test_cases:
    run_test_case(case_number, input_nums, expected)
```
