# **SOLUTION**

```python
from collections import deque
class Solution(object):
    def findLexSmallestString(self, s, a, b):
        """
        :type s: str
        :type a: int
        :type b: int
        :rtype: str
        """
        return self.solverBFS(s, a, b)

    def operationAdd(self, s, a):
        chars = list(s)
        n = len(chars)
        for i in range(1, n, 2):
            chars[i] = str((int(chars[i]) + a) % 10)

        return "".join(chars)
    
    def operationRot(self, s, b):
        return s[-b:] + s[:-b]
    
    def solverBFS(self, s, a, b):
        visited = set()
        queue = deque([s])
        smallest = s

        while queue:
            cur = queue.popleft()

            if cur in visited:
                continue
            
            queue.append(self.operationAdd(cur, a))
            queue.append(self.operationRot(cur, b))
            visited.add(cur)

            smallest = min(smallest, cur)

        return smallest

        
```
