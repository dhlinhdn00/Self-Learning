```python
import sys

def solve():
    data = sys.stdin.read().strip().split()
    it = iter(data)
    t = int(next(it))
    out_lines = []
    for _ in range(t):
        n = int(next(it))
        arr = [int(next(it)) for _ in range(n)]

        first = None
        last = None
        for i, v in enumerate(arr):
            if v != i + 1 or v == 0:
                if first is None:
                    first = i
                last = i

        ans = last - first + 1 
        out_lines.append(str(ans))
    sys.stdout.write("\n".join(out_lines))

if __name__ == "__main__":
    solve()
```
