# **SOLUTION**

```python
def solver(n, s):
    if all(s[i] <= s[i+1] for i in range(n-1)):
        return 0
    
    count_10 = sum(1 for i in range(n-1) if s[i] == '1' and s[i+1] == '0')
    
    if count_10 == 1:
        return 1
    else:
        return 2


def solution():
    t = int(input())
    for _ in range(t):
        n = int(input())
        s = input().strip()
        print(solver(n, s))
```
